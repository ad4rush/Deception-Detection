```python
# %%
# Import necessary libraries
import pandas as pd
import numpy as np
import json
import os
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
import re
import string
import glob # For finding files
import warnings
import random

# Scikit-learn imports
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix
from sklearn.utils import resample

# TensorFlow / Keras imports
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import (
    Input, Embedding, Dense, Dropout, Flatten, Concatenate,
    Layer, MultiHeadAttention, LayerNormalization, GlobalAveragePooling1D, Add
)
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras import backend as K

# Filter out specific warnings if they are noisy and understood
warnings.filterwarnings("ignore", category=UserWarning, module='tensorflow')

# Set random seeds for reproducibility
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

# --- Constants and Configuration ---
BASE_DATA_DIR = '/kaggle/input/diplomacy/transformers/default/1/Dataset/'
DATA_DIR = os.path.join(BASE_DATA_DIR, 'data')
MOVES_DIR = os.path.join(BASE_DATA_DIR, 'moves')
OUTPUT_DIR = '/kaggle/working/'

# Model Hyperparameters
MAX_VOCAB_SIZE = 10000
MAX_SEQUENCE_LENGTH = 60
EMBEDDING_DIM = 128
NUM_HEADS = 4
DROPOUT_RATE = 0.1
# Training Params
BATCH_SIZE = 32
EPOCHS = 25
PATIENCE = 5
LEARNING_RATE = 1e-4
BALANCE_METHOD = 'downsample'

os.makedirs(OUTPUT_DIR, exist_ok=True)

# --- Territory/Keyword Definitions ---
# (Same as your previous version)
DIPLOMACY_TERRITORIES=sorted(list(set(['adr','aeg','alb','ank','apu','arm','bal','bar','bel','ber','bla','boh','bre','bud','bul','bur','cly','con','den','eas','edi','eng','fin','gal','gas','gre','gulf','hel','hol','ion','iri','kie','lon','lvp','lvn','mao','mar','mid','mos','mun','naf','nap','nrg','nth','nwy','par','pic','pie','por','pru','rom','ruh','rum','ser','sev','ska','sil','smy','spa','st','swe','syr','tri','tun','tus','tyr','tys','ukr','ven','vie','wal','war','wes','yor','stp',])))
TERRITORY_REGEX=r'\b('+'|'.join(DIPLOMACY_TERRITORIES)+r')\b'; SUPPORT_KEYWORDS=['support','help','assist','back','aid']
ATTACK_MOVE_KEYWORDS=['attack','move','take','go','advance','invade','push','head','walk','march','grab','occupy']
HOLD_KEYWORDS=['hold','defend','stay','keep','protect','guard','remain']; SUCCESS_KEYWORDS=['succeed','win','works','successful','guarantee','sure','make it','get it','certain']
FAIL_KEYWORDS=['fail','lose','won\'t work','wont work','doubtful','risky','might not','can\'t','cannot','difficult','unlikely']

# --- Helper Functions ---

def clean_text(text):
    if not isinstance(text, str): return ""
    text = text.lower(); text = re.sub(r'http\S+', '', text); text = re.sub(r'@\w+', '', text); text = re.sub(r'#\w+', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation.replace("'", ""))); text = re.sub(r'\d+', '', text); text = re.sub(r'\s+', ' ', text).strip()
    return text

def load_jsonl_dataset(file_path):
    data = []; skipped_lines = 0
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            for i, line in enumerate(f):
                try: data.append(json.loads(line))
                except json.JSONDecodeError: skipped_lines += 1; # print(f"Warn: Skip JSON line {i+1} in {file_path}")
    except FileNotFoundError: print(f"Error: File not found at {file_path}"); return []
    if skipped_lines > 0: print(f"Warning: Skipped {skipped_lines} invalid JSON lines in {file_path}")
    return data

def load_moves_data(moves_dir):
    moves_lookup = {}; print(f"Loading moves data from: {moves_dir}")
    if not os.path.isdir(moves_dir): print(f"Error: Moves dir not found: {moves_dir}"); return {}
    loaded_count=0; file_count=0; error_files=0
    for filepath in glob.glob(os.path.join(moves_dir, '*.json')):
        file_count+=1; filename=os.path.basename(filepath); match=re.match(r'DiplomacyGame(\d+)_(\d+)_(\w+)\.json', filename)
        if match:
            try: game_id=int(match.group(1)); year=match.group(2); season=match.group(3).lower();
            except ValueError: # print(f"Warn: Parse ID/Year fail {filename}");
                error_files+=1; continue
            try:
                with open(filepath,'r',encoding='utf-8') as f: moves_lookup[(game_id,year,season)]=json.load(f); loaded_count+=1
            except Exception as e: # print(f"Warn: Load fail {filename}: {e}");
                 error_files+=1
    print(f"Loaded data for {loaded_count} game turns (out of {file_count} files). {error_files} files had errors.")
    return moves_lookup

# --- Preprocessing for Exploration and Modeling ---
def preprocess_data_for_exploration_and_model(data_raw):
    all_processed_samples = [] # Keep all for initial stats
    annotated_samples_for_model = [] # Filtered for modeling
    no_annotation_samples = []
    skipped_dialogues = 0; total_msgs = 0

    for dialogue_idx, dialogue_sample in enumerate(data_raw):
        messages=dialogue_sample.get('messages',[]); sender_labels=dialogue_sample.get('sender_labels',[]); receiver_labels=dialogue_sample.get('receiver_labels',[])
        speakers=dialogue_sample.get('speakers',[]); recipients=dialogue_sample.get('receivers',[]); game_id=dialogue_sample.get('game_id')
        years=dialogue_sample.get('years',[]); seasons=dialogue_sample.get('seasons',[]); game_scores=dialogue_sample.get('game_score',[])
        game_score_deltas=dialogue_sample.get('game_score_delta',[])
        list_len=len(messages); total_msgs += list_len
        req_lengths=[sender_labels,receiver_labels,speakers,recipients,years,seasons,game_scores,game_score_deltas]
        if not (game_id and list_len>0 and all(len(lst)==list_len for lst in req_lengths)): skipped_dialogues+=1; continue

        for i in range(list_len):
            delta_num = 0.0 # Default delta feature
            delta_str = game_score_deltas[i] if game_score_deltas[i] else 'N/A'
            try: delta_num = float(delta_str)
            except (ValueError, TypeError): delta_str = 'N/A' # Ensure it stays N/A if conversion fails

            sample = {
                 'text': messages[i] or "", 'clean_text': clean_text(messages[i] or ""),
                 'game_id': game_id, 'speaker': str(speakers[i] or 'unk').lower(), 'recipient': str(recipients[i] or 'unk').lower(),
                 'year': str(years[i] or 'unk'), 'season': str(seasons[i] or 'unk').lower(),
                 'score': game_scores[i] or 'N/A', 'delta': delta_str, 'delta_numeric': delta_num, # Keep both str & numeric
                 'receiver_annotation': receiver_labels[i], 'sender_label_raw': sender_labels[i] # Keep raw sender label
            }

            if sender_labels[i] is not None: sample['label'] = 0 if sender_labels[i] == False else 1
            else: sample['label'] = -1 # Indicate missing/invalid sender label

            all_processed_samples.append(sample)
            if sample['receiver_annotation'] == "NOANNOTATION": no_annotation_samples.append(sample)
            elif sample['label'] != -1 : annotated_samples_for_model.append(sample) # Keep only valid annotated for model

    print(f"Preprocessed {len(all_processed_samples)} total messages from {dialogue_idx+1-skipped_dialogues} valid dialogues.")
    if skipped_dialogues > 0: print(f"  Warning: Skipped {skipped_dialogues} dialogues due to inconsistent list lengths or missing game_id.")
    print(f"  Found {len(no_annotation_samples)} NOANNOTATION receiver labels.")
    print(f"  Found {len(annotated_samples_for_model)} samples with valid annotations for modeling.")
    return annotated_samples_for_model, no_annotation_samples, all_processed_samples

# Feature Extraction function (appends delta_numeric)
def extract_action_features(sample_info, moves_lookup):
    text=sample_info.get('clean_text',''); speaker=sample_info.get('speaker',''); game_id=sample_info.get('game_id')
    year=sample_info.get('year'); season=sample_info.get('season');
    # --- Start with action/keyword features ---
    # (Same keyword/action logic as previous version)
    default_len=13; default_features=[0.0]*default_len;
    if not all([game_id,year,season,speaker]): action_features = default_features # Keep default if no context
    else: # Proceed with action feature extraction
        move_key=(game_id,year,season); move_info=moves_lookup.get(move_key);
        if move_info is None or 'orders' not in move_info: action_features = default_features # Keep default if no moves data
        else: # Actual feature calculation (condensed from previous)
            mentioned_support_keyword=1.0 if any(kw in text for kw in SUPPORT_KEYWORDS) else 0.0; mentioned_attack_move=1.0 if any(kw in text for kw in ATTACK_MOVE_KEYWORDS) else 0.0
            mentioned_hold=1.0 if any(kw in text for kw in HOLD_KEYWORDS) else 0.0; claim_success=1.0 if any(kw in text for kw in SUCCESS_KEYWORDS) else 0.0
            claim_fail=1.0 if any(kw in text for kw in FAIL_KEYWORDS) else 0.0; gave_support_order,gave_move_order,gave_hold_order=0.0,0.0,0.0
            promised_support_given,promised_attack_executed,promised_hold_kept=0.0,0.0,0.0; statement_contradicts_move,move_success_claim_match=0.0,0.0
            mentioned_territories=re.findall(TERRITORY_REGEX,text); first_mentioned_territory=mentioned_territories[0] if mentioned_territories else None
            actual_order_for_territory=None; speaker_orders_dict={}
            speaker_orders_raw=move_info.get('orders',{});
            for k in speaker_orders_raw:
                if k.lower()==speaker: speaker_orders_dict=speaker_orders_raw[k]; break
            if speaker_orders_dict:
                for unit_loc_key,order_details in speaker_orders_dict.items():
                    if not isinstance(order_details,dict): continue
                    unit_loc=unit_loc_key.lower(); order_type=order_details.get('type','').upper()
                    if order_type=='SUPPORT': gave_support_order=1.0
                    elif order_type=='MOVE': gave_move_order=1.0
                    elif order_type=='HOLD': gave_hold_order=1.0
                    if first_mentioned_territory:
                        involved_territories={unit_loc}; to_loc=order_details.get('to',''); from_loc=order_details.get('from','')
                        if to_loc: involved_territories.add(str(to_loc).lower())
                        if order_type=='SUPPORT' and from_loc: involved_territories.add(str(from_loc).lower())
                        if first_mentioned_territory in involved_territories:
                            if unit_loc==first_mentioned_territory or actual_order_for_territory is None: actual_order_for_territory=order_details
            if actual_order_for_territory:
                order_type=actual_order_for_territory.get('type','').upper(); order_result=actual_order_for_territory.get('result','').upper()
                if mentioned_support_keyword and order_type=='SUPPORT': promised_support_given=1.0
                if mentioned_attack_move and order_type=='MOVE': promised_attack_executed=1.0
                if mentioned_hold and order_type=='HOLD': promised_hold_kept=1.0
                if mentioned_attack_move and order_type=='SUPPORT': statement_contradicts_move=1.0
                if order_result in ['SUCCEEDS','FAILS']:
                    if claim_success and order_result=='SUCCEEDS': move_success_claim_match=1.0
                    elif claim_fail and order_result=='FAILS': move_success_claim_match=1.0
                    elif (claim_success and order_result=='FAILS') or (claim_fail and order_result=='SUCCEEDS'): move_success_claim_match=-1.0
            action_features=[mentioned_support_keyword,gave_support_order,mentioned_attack_move,gave_move_order,mentioned_hold,gave_hold_order,
                             promised_support_given,promised_attack_executed,promised_hold_kept,statement_contradicts_move,claim_success,claim_fail,move_success_claim_match]

    # --- Append the numeric delta score ---
    delta_numeric = sample_info.get('delta_numeric', 0.0) # Get the pre-calculated numeric delta
    action_features.append(delta_numeric) # Add it as the last feature

    return action_features

# Function for Delta Score Analysis
def analyze_delta_scores(samples_list, label_name):
    deltas = []
    invalid_count = 0
    for sample in samples_list:
        delta_str = sample.get('delta', 'N/A')
        try: deltas.append(float(delta_str))
        except (ValueError, TypeError): invalid_count += 1
    total_count = len(samples_list)
    valid_count = len(deltas)
    if total_count==0: print(f"No {label_name} samples."); return
    neg_delta = sum(1 for d in deltas if d<0); zero_delta=sum(1 for d in deltas if d==0); pos_delta=sum(1 for d in deltas if d>0)
    print(f"\n--- Delta Score Analysis for {label_name} (Total Valid: {valid_count} / {total_count}) ---")
    if valid_count > 0:
      print(f"  Negative (< 0): {neg_delta} ({neg_delta/valid_count:.1%})")
      print(f"  Zero     (= 0): {zero_delta} ({zero_delta/valid_count:.1%})")
      print(f"  Positive (> 0): {pos_delta} ({pos_delta/valid_count:.1%})")
      print(f"  Non-Neg (>= 0): {zero_delta + pos_delta} ({(zero_delta + pos_delta)/valid_count:.1%})")
    if invalid_count>0: print(f"  Invalid/NA    : {invalid_count} ({invalid_count/total_count:.1%})")

# --- Main Script Logic ---

print("--- Starting Data Loading ---")
train_data_raw = load_jsonl_dataset(os.path.join(DATA_DIR, 'train.jsonl'))
val_data_raw = load_jsonl_dataset(os.path.join(DATA_DIR, 'validation.jsonl'))
test_data_raw = load_jsonl_dataset(os.path.join(DATA_DIR, 'test.jsonl'))
moves_lookup = load_moves_data(MOVES_DIR)
if not train_data_raw or not val_data_raw or not test_data_raw: raise ValueError("Dataset loading failed.")

# --- Data Exploration & Preparation ---
print("\n--- Data Exploration & Preparation ---")
train_model_samples, train_no_annotation, train_all = preprocess_data_for_exploration_and_model(train_data_raw)
val_model_samples, _, _ = preprocess_data_for_exploration_and_model(val_data_raw)
test_model_samples, _, _ = preprocess_data_for_exploration_and_model(test_data_raw)

lies = [s for s in train_model_samples if s['label']==0]; truths = [s for s in train_model_samples if s['label']==1]
print(f"\nUsing {len(train_model_samples)} annotated samples for training/validation.")
num_examples_to_show = 5 # Reduced example print count
# Print Lie/Truth/NoAnn Examples (Concise format)
print(f"\n--- {num_examples_to_show} Lie Examples (Train Set) ---")
if lies:
    for i,s in enumerate(random.sample(lies,min(num_examples_to_show,len(lies)))):print(f"L{i+1} Δ:{s['delta']:<4} | Txt:'{s['text'][:70]}...'")
else: print("No lie examples.")
print(f"\n--- {num_examples_to_show} Truth Examples (Train Set) ---")
if truths:
    for i,s in enumerate(random.sample(truths,min(num_examples_to_show,len(truths)))):print(f"T{i+1} Δ:{s['delta']:<4} | Txt:'{s['text'][:70]}...'")
else: print("No truth examples.")
print(f"\n--- {num_examples_to_show} NOANNOTATION Examples (Train Set) ---")
if train_no_annotation:
    for i,s in enumerate(random.sample(train_no_annotation,min(num_examples_to_show,len(train_no_annotation)))): print(f"N{i+1} Sender:{'L' if s['label']==0 else 'T' if s['label']==1 else '?'} | Txt:'{s['text'][:70]}...'")
else: print("No NOANNOTATION examples.")
# Analyze Delta Scores
analyze_delta_scores(lies, "Lies (Train Annotated)")
analyze_delta_scores(truths, "Truths (Train Annotated)")

# --- Final Data Prep for Model ---
train_texts_clean=[s['clean_text'] for s in train_model_samples]; train_features=np.array([extract_action_features(s,moves_lookup) for s in train_model_samples]); train_labels=np.array([s['label'] for s in train_model_samples])
val_texts_clean=[s['clean_text'] for s in val_model_samples]; val_features=np.array([extract_action_features(s,moves_lookup) for s in val_model_samples]); val_labels=np.array([s['label'] for s in val_model_samples])
test_texts_clean=[s['clean_text'] for s in test_model_samples]; test_features=np.array([extract_action_features(s,moves_lookup) for s in test_model_samples]); test_labels=np.array([s['label'] for s in test_model_samples])

if len(train_labels)==0 or len(val_labels)==0 or len(test_labels)==0: raise ValueError("Modeling data splits empty.")
print("\nShapes of data prepared for modeling:")
print(f"Train: texts={len(train_texts_clean)}, feats={train_features.shape}, labels={train_labels.shape}")
print(f"Val:   texts={len(val_texts_clean)}, feats={val_features.shape}, labels={val_labels.shape}")
print(f"Test:  texts={len(test_texts_clean)}, feats={test_features.shape}, labels={test_labels.shape}")
num_features = train_features.shape[1]; print(f"Number of features per sample (incl. delta): {num_features}")

# --- Tokenization and Padding ---
print("\nTokenizing and padding text..."); tokenizer=Tokenizer(num_words=MAX_VOCAB_SIZE, oov_token='<OOV>')
tokenizer.fit_on_texts(train_texts_clean); vocab_size=min(MAX_VOCAB_SIZE,len(tokenizer.word_index)+1); print(f"Vocab size:{vocab_size}")
train_sequences=tokenizer.texts_to_sequences(train_texts_clean); val_sequences=tokenizer.texts_to_sequences(val_texts_clean); test_sequences=tokenizer.texts_to_sequences(test_texts_clean)
train_padded=pad_sequences(train_sequences,maxlen=MAX_SEQUENCE_LENGTH,padding='post',truncating='post')
val_padded=pad_sequences(val_sequences,maxlen=MAX_SEQUENCE_LENGTH,padding='post',truncating='post'); test_padded=pad_sequences(test_sequences,maxlen=MAX_SEQUENCE_LENGTH,padding='post',truncating='post')
print(f"Padded shapes: Train:{train_padded.shape}, Val:{val_padded.shape}, Test:{test_padded.shape}")

# --- Data Balancing ---
if BALANCE_METHOD=='downsample':
    print("\nBalancing training data (downsample)..."); train_df=pd.DataFrame({'pad':list(train_padded),'feat':list(train_features),'lbl':train_labels})
    df_maj=train_df[train_df.lbl==1]; df_min=train_df[train_df.lbl==0]; print(f"Orig counts: Truth(1):{len(df_maj)}, Deceptive(0):{len(df_min)}")
    if len(df_min)==0: print("Warn: Minority empty."); bal_train_pad,bal_train_feat,bal_train_lbl=train_padded,train_features,train_labels
    else:
        df_maj_ds=resample(df_maj,replace=False,n_samples=len(df_min),random_state=SEED); df_bal=pd.concat([df_min,df_maj_ds]).sample(frac=1,random_state=SEED).reset_index(drop=True)
        bal_train_pad=np.array(df_bal['pad'].tolist()); bal_train_feat=np.array(df_bal['feat'].tolist()); bal_train_lbl=np.array(df_bal['lbl'].tolist())
        print(f"Balanced counts: Truth(1):{sum(bal_train_lbl==1)}, Deceptive(0):{sum(bal_train_lbl==0)}")
    print(f"Shapes after balancing: padded={bal_train_pad.shape}, feats={bal_train_feat.shape}, labels={bal_train_lbl.shape}")
else: print("\nNo balancing."); bal_train_pad,bal_train_feat,bal_train_lbl=train_padded,train_features,train_labels
if len(bal_train_lbl)==0: raise ValueError("Train data empty after balancing.")

# --- Model Definition (Simple Transformer + Features + Delta) ---
def create_simple_transformer_model_with_features(vocab_size, embedding_dim, num_heads, max_sequence_length, num_total_features, dropout_rate=0.1): # Use num_total_features
    text_input=Input(shape=(max_sequence_length,),name='text_input'); embedding=Embedding(input_dim=vocab_size,output_dim=embedding_dim)(text_input)
    attn_output=MultiHeadAttention(num_heads=num_heads,key_dim=embedding_dim)(query=embedding,value=embedding,key=embedding)
    attn_output=Dropout(dropout_rate)(attn_output); out1=LayerNormalization(epsilon=1e-6)(Add()([embedding,attn_output]))
    text_branch=GlobalAveragePooling1D()(out1); text_branch=Dropout(0.4)(text_branch)
    # Input shape uses num_total_features now
    feature_input=Input(shape=(num_total_features,),name='feature_input')
    concatenated=Concatenate()([text_branch,feature_input])
    dense1=Dense(64,activation='relu')(concatenated); dropout1=Dropout(0.5)(dense1); outputs=Dense(1,activation='sigmoid',name='output')(dropout1)
    model=Model(inputs=[text_input,feature_input],outputs=outputs,name="SimpleTransformer_plus_Features_and_Delta")
    return model

# --- Compile and Train ---
print("\n--- Training Simple Transformer + Action Features + Delta (Downsampled) ---")
model_name = "SimpleTransformer_plus_Features_Delta_Downsampled"
# Ensure num_features used in model definition includes the added delta feature
model = create_simple_transformer_model_with_features(
    vocab_size, EMBEDDING_DIM, NUM_HEADS, MAX_SEQUENCE_LENGTH, num_features, DROPOUT_RATE
)
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE),loss='binary_crossentropy',metrics=['accuracy'])
model.summary()

model_filename = model_name.lower() + ".keras"; model_path = os.path.join(OUTPUT_DIR, f"best_{model_filename}")
early_stopping = EarlyStopping(monitor='val_loss',patience=PATIENCE,verbose=1,restore_best_weights=True)
model_checkpoint = ModelCheckpoint(model_path,monitor='val_loss',save_best_only=True,mode='min',verbose=1)
print(f"\nTraining {model_name}...")
history=model.fit([bal_train_pad,bal_train_feat],bal_train_lbl,batch_size=BATCH_SIZE,epochs=EPOCHS,
                  validation_data=([val_padded,val_features],val_labels),callbacks=[early_stopping,model_checkpoint],verbose=1)

# --- Evaluation ---
print(f"\n--- Evaluating Best {model_name} on Test Set ---")
print(f"Loading best weights from {model_path}")
try: model.load_weights(model_path)
except Exception as e: print(f"Warn: Load weights fail {model_path}: {e}")

test_inputs = [test_padded,test_features]; test_loss,test_accuracy = model.evaluate(test_inputs,test_labels,verbose=0)
test_pred_probs = model.predict(test_inputs); test_pred_bin = (test_pred_probs > 0.5).astype(int)

prec=precision_score(test_labels,test_pred_bin,pos_label=1,zero_division=0); rec=recall_score(test_labels,test_pred_bin,pos_label=1,zero_division=0)
f1=f1_score(test_labels,test_pred_bin,pos_label=1,zero_division=0); f1_macro=f1_score(test_labels,test_pred_bin,average='macro',zero_division=0)
f1_lie=f1_score(test_labels,test_pred_bin,pos_label=0,average='binary',zero_division=0)
results={model_name:{"Accuracy":test_accuracy,"Macro F1":f1_macro,"Lie F1 (0)":f1_lie,"F1 (1)":f1,"Prec (1)":prec,"Rec (1)":rec,"Loss":test_loss}}
print(f"\n[{model_name}] Test Set Metrics:"); print(f" Accuracy:{test_accuracy:.4f}, MacroF1:{f1_macro:.4f}, LieF1:{f1_lie:.4f}, Loss:{test_loss:.4f}")

conf_matrix=confusion_matrix(test_labels,test_pred_bin); plt.figure(figsize=(7,5));
sns.heatmap(conf_matrix,annot=True,fmt='d',cmap='Blues',xticklabels=['Pred Decep(0)','Pred Truth(1)'],yticklabels=['Actual Decep(0)','Actual Truth(1)'])
plt.title(f'[{model_name}]\nMacro F1:{f1_macro:.4f}|Lie F1:{f1_lie:.4f}',fontsize=10); plt.xlabel('Predicted'); plt.ylabel('Actual'); plt.tight_layout()
plot_path=os.path.join(OUTPUT_DIR,f"cm_{model_filename.replace('.keras','.png')}"); plt.savefig(plot_path); print(f"Saved CM to {plot_path}"); plt.show()

print("\n--- Final Model Summary (Test Set) ---")
if results: results_df=pd.DataFrame(results).T; cols_order=["Accuracy","Macro F1","Lie F1 (0)","F1 (1)","Prec (1)","Rec (1)","Loss","Error"]; results_df=results_df.reindex(columns=cols_order); print(results_df.to_markdown(floatfmt=".4f"))
else: print("No models eval.")
print("\n--- Script Finished ---")
```

```python
# %% [code]
# Import necessary libraries
import pandas as pd
import numpy as np
import json
import os
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
import re
import string
import glob  # For finding files
import warnings
import random

# Scikit-learn imports
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix
from sklearn.utils import resample

# TensorFlow / Keras imports
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import (
    Input, Embedding, Dense, Dropout, Flatten, Concatenate,
    Layer, MultiHeadAttention, LayerNormalization, GlobalAveragePooling1D, Add
)
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras import backend as K

# Filter out specific warnings if they are noisy and understood
warnings.filterwarnings("ignore", category=UserWarning, module='tensorflow')

# Set random seeds for reproducibility
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

# --- Constants and Configuration ---
BASE_DATA_DIR = '/kaggle/input/diplomacy/transformers/default/1/Dataset/'
DATA_DIR = os.path.join(BASE_DATA_DIR, 'data')
MOVES_DIR = os.path.join(BASE_DATA_DIR, 'moves')
OUTPUT_DIR = '/kaggle/working/'

# Model Hyperparameters
MAX_VOCAB_SIZE = 10000
MAX_SEQUENCE_LENGTH = 60
EMBEDDING_DIM = 128
NUM_HEADS = 4
DROPOUT_RATE = 0.1
# Training Params
BATCH_SIZE = 32
EPOCHS = 25
PATIENCE = 5
LEARNING_RATE = 1e-4
BALANCE_METHOD = 'downsample'

os.makedirs(OUTPUT_DIR, exist_ok=True)

# --- Territory/Keyword Definitions ---
DIPLOMACY_TERRITORIES = sorted(list(set([
    'adr','aeg','alb','ank','apu','arm','bal','bar','bel','ber','bla','boh','bre','bud','bul',
    'bur','cly','con','den','eas','edi','eng','fin','gal','gas','gre','gulf','hel','hol','ion',
    'iri','kie','lon','lvp','lvn','mao','mar','mid','mos','mun','naf','nap','nrg','nth','nwy',
    'par','pic','pie','por','pru','rom','ruh','rum','ser','sev','ska','sil','smy','spa','st','swe',
    'syr','tri','tun','tus','tyr','tys','ukr','ven','vie','wal','war','wes','yor','stp',
])))
TERRITORY_REGEX = r'\b(' + '|'.join(DIPLOMACY_TERRITORIES) + r')\b'
SUPPORT_KEYWORDS = ['support','help','assist','back','aid']
ATTACK_MOVE_KEYWORDS = ['attack','move','take','go','advance','invade','push','head','walk','march','grab','occupy']
HOLD_KEYWORDS = ['hold','defend','stay','keep','protect','guard','remain']
SUCCESS_KEYWORDS = ['succeed','win','works','successful','guarantee','sure','make it','get it','certain']
FAIL_KEYWORDS = ['fail','lose','won\'t work','wont work','doubtful','risky','might not','can\'t','cannot','difficult','unlikely']

# --- Helper Functions ---

def clean_text(text):
    if not isinstance(text, str): 
        return ""
    text = text.lower()
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'#\w+', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation.replace("'", "")))
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def load_jsonl_dataset(file_path):
    data = []
    skipped_lines = 0
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            for i, line in enumerate(f):
                try: 
                    data.append(json.loads(line))
                except json.JSONDecodeError:
                    skipped_lines += 1
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
        return []
    if skipped_lines > 0: 
        print(f"Warning: Skipped {skipped_lines} invalid JSON lines in {file_path}")
    return data

def load_moves_data(moves_dir):
    moves_lookup = {}
    print(f"Loading moves data from: {moves_dir}")
    if not os.path.isdir(moves_dir):
        print(f"Error: Moves dir not found: {moves_dir}")
        return {}
    loaded_count = 0
    file_count = 0
    error_files = 0
    for filepath in glob.glob(os.path.join(moves_dir, '*.json')):
        file_count += 1
        filename = os.path.basename(filepath)
        match = re.match(r'DiplomacyGame(\d+)_(\d+)_(\w+)\.json', filename)
        if match:
            try:
                game_id = int(match.group(1))
                year = match.group(2)
                season = match.group(3).lower()
            except ValueError:
                error_files += 1
                continue
            try:
                with open(filepath, 'r', encoding='utf-8') as f:
                    moves_lookup[(game_id, year, season)] = json.load(f)
                loaded_count += 1
            except Exception as e:
                error_files += 1
    print(f"Loaded data for {loaded_count} game turns (out of {file_count} files). {error_files} files had errors.")
    return moves_lookup

# --- Modified Preprocessing Function ---
def preprocess_data_for_exploration_and_model(data_raw):
    all_processed_samples = []  # Keep all for initial stats
    annotated_samples_for_model = []  # Filtered for modeling
    no_annotation_samples = []
    skipped_dialogues = 0
    total_msgs = 0

    for dialogue_idx, dialogue_sample in enumerate(data_raw):
        messages = dialogue_sample.get('messages', [])
        sender_labels = dialogue_sample.get('sender_labels', [])
        receiver_labels = dialogue_sample.get('receiver_labels', [])
        speakers = dialogue_sample.get('speakers', [])
        recipients = dialogue_sample.get('receivers', [])
        game_id = dialogue_sample.get('game_id')
        years = dialogue_sample.get('years', [])
        seasons = dialogue_sample.get('seasons', [])
        game_scores = dialogue_sample.get('game_score', [])
        game_score_deltas = dialogue_sample.get('game_score_delta', [])
        list_len = len(messages)
        total_msgs += list_len
        req_lengths = [sender_labels, receiver_labels, speakers, recipients, years, seasons, game_scores, game_score_deltas]
        if not (game_id and list_len > 0 and all(len(lst) == list_len for lst in req_lengths)):
            skipped_dialogues += 1
            continue

        for i in range(list_len):
            delta_str = game_score_deltas[i] if game_score_deltas[i] else 'N/A'
            try:
                delta_num = float(delta_str)
            except (ValueError, TypeError):
                delta_num = 0.0
                delta_str = 'N/A'
            sample = {
                 'text': messages[i] or "",
                 'clean_text': clean_text(messages[i] or ""),
                 'game_id': game_id,
                 'speaker': str(speakers[i] or 'unk').lower(),
                 'recipient': str(recipients[i] or 'unk').lower(),
                 'year': str(years[i] or 'unk'),
                 'season': str(seasons[i] or 'unk').lower(),
                 'score': game_scores[i] or 'N/A',
                 'delta': delta_str,
                 'delta_numeric': delta_num,
                 'receiver_annotation': receiver_labels[i],
                 'sender_label_raw': sender_labels[i],
                 'relative_index': i,
                 'dialogue_length': list_len
            }
            # Determine the relative instance: past, present, or future
            if list_len > 0:
                if i < list_len / 3:
                    sample['relative_instance'] = 'past'
                elif i < 2 * list_len / 3:
                    sample['relative_instance'] = 'present'
                else:
                    sample['relative_instance'] = 'future'
            else:
                sample['relative_instance'] = 'unknown'
            
            all_processed_samples.append(sample)
            if sample['receiver_annotation'] == "NOANNOTATION":
                no_annotation_samples.append(sample)
            elif sample['sender_label_raw'] is not None:
                # Assign label: 0 for deceptive (False) and 1 for truthful (True)
                sample['label'] = 0 if sample['sender_label_raw'] == False else 1
                annotated_samples_for_model.append(sample)
            else:
                sample['label'] = -1  # Indicate missing/invalid sender label

    print(f"Preprocessed {len(all_processed_samples)} total messages from {dialogue_idx + 1 - skipped_dialogues} valid dialogues.")
    if skipped_dialogues > 0:
        print(f"  Warning: Skipped {skipped_dialogues} dialogues due to inconsistent list lengths or missing game_id.")
    print(f"  Found {len(no_annotation_samples)} NOANNOTATION receiver labels.")
    print(f"  Found {len(annotated_samples_for_model)} samples with valid annotations for modeling.")
    return annotated_samples_for_model, no_annotation_samples, all_processed_samples

# Feature Extraction function (appends delta_numeric)
def extract_action_features(sample_info, moves_lookup):
    text = sample_info.get('clean_text', '')
    speaker = sample_info.get('speaker', '')
    game_id = sample_info.get('game_id')
    year = sample_info.get('year')
    season = sample_info.get('season')
    default_len = 13
    default_features = [0.0] * default_len
    if not all([game_id, year, season, speaker]):
        action_features = default_features
    else:
        move_key = (game_id, year, season)
        move_info = moves_lookup.get(move_key)
        if move_info is None or 'orders' not in move_info:
            action_features = default_features
        else:
            mentioned_support_keyword = 1.0 if any(kw in text for kw in SUPPORT_KEYWORDS) else 0.0
            mentioned_attack_move = 1.0 if any(kw in text for kw in ATTACK_MOVE_KEYWORDS) else 0.0
            mentioned_hold = 1.0 if any(kw in text for kw in HOLD_KEYWORDS) else 0.0
            claim_success = 1.0 if any(kw in text for kw in SUCCESS_KEYWORDS) else 0.0
            claim_fail = 1.0 if any(kw in text for kw in FAIL_KEYWORDS) else 0.0
            gave_support_order = 0.0
            gave_move_order = 0.0
            gave_hold_order = 0.0
            promised_support_given = 0.0
            promised_attack_executed = 0.0
            promised_hold_kept = 0.0
            statement_contradicts_move = 0.0
            move_success_claim_match = 0.0
            mentioned_territories = re.findall(TERRITORY_REGEX, text)
            first_mentioned_territory = mentioned_territories[0] if mentioned_territories else None
            actual_order_for_territory = None
            speaker_orders_dict = {}
            speaker_orders_raw = move_info.get('orders', {})
            for k in speaker_orders_raw:
                if k.lower() == speaker:
                    speaker_orders_dict = speaker_orders_raw[k]
                    break
            if speaker_orders_dict:
                for unit_loc_key, order_details in speaker_orders_dict.items():
                    if not isinstance(order_details, dict):
                        continue
                    unit_loc = unit_loc_key.lower()
                    order_type = order_details.get('type', '').upper()
                    if order_type == 'SUPPORT':
                        gave_support_order = 1.0
                    elif order_type == 'MOVE':
                        gave_move_order = 1.0
                    elif order_type == 'HOLD':
                        gave_hold_order = 1.0
                    if first_mentioned_territory:
                        involved_territories = {unit_loc}
                        to_loc = order_details.get('to', '')
                        from_loc = order_details.get('from', '')
                        if to_loc:
                            involved_territories.add(str(to_loc).lower())
                        if order_type == 'SUPPORT' and from_loc:
                            involved_territories.add(str(from_loc).lower())
                        if first_mentioned_territory in involved_territories:
                            if unit_loc == first_mentioned_territory or actual_order_for_territory is None:
                                actual_order_for_territory = order_details
            if actual_order_for_territory:
                order_type = actual_order_for_territory.get('type', '').upper()
                order_result = actual_order_for_territory.get('result', '').upper()
                if mentioned_support_keyword and order_type == 'SUPPORT':
                    promised_support_given = 1.0
                if mentioned_attack_move and order_type == 'MOVE':
                    promised_attack_executed = 1.0
                if mentioned_hold and order_type == 'HOLD':
                    promised_hold_kept = 1.0
                if mentioned_attack_move and order_type == 'SUPPORT':
                    statement_contradicts_move = 1.0
                if order_result in ['SUCCEEDS','FAILS']:
                    if claim_success and order_result == 'SUCCEEDS':
                        move_success_claim_match = 1.0
                    elif claim_fail and order_result == 'FAILS':
                        move_success_claim_match = 1.0
                    elif (claim_success and order_result == 'FAILS') or (claim_fail and order_result == 'SUCCEEDS'):
                        move_success_claim_match = -1.0
            action_features = [mentioned_support_keyword, gave_support_order, mentioned_attack_move, gave_move_order, mentioned_hold, gave_hold_order,
                               promised_support_given, promised_attack_executed, promised_hold_kept, statement_contradicts_move, claim_success, claim_fail, move_success_claim_match]

    # Append the numeric delta score as the last feature
    delta_numeric = sample_info.get('delta_numeric', 0.0)
    action_features.append(delta_numeric)
    return action_features

# Function for Delta Score Analysis (existing)
def analyze_delta_scores(samples_list, label_name):
    deltas = []
    invalid_count = 0
    for sample in samples_list:
        delta_str = sample.get('delta', 'N/A')
        try:
            deltas.append(float(delta_str))
        except (ValueError, TypeError):
            invalid_count += 1
    total_count = len(samples_list)
    valid_count = len(deltas)
    if total_count == 0:
        print(f"No {label_name} samples.")
        return
    neg_delta = sum(1 for d in deltas if d < 0)
    zero_delta = sum(1 for d in deltas if d == 0)
    pos_delta = sum(1 for d in deltas if d > 0)
    print(f"\n--- Delta Score Analysis for {label_name} (Total Valid: {valid_count} / {total_count}) ---")
    if valid_count > 0:
        print(f"  Negative (< 0): {neg_delta} ({neg_delta/valid_count:.1%})")
        print(f"  Zero     (= 0): {zero_delta} ({zero_delta/valid_count:.1%})")
        print(f"  Positive (> 0): {pos_delta} ({pos_delta/valid_count:.1%})")
        print(f"  Non-Neg (>= 0): {zero_delta + pos_delta} ({(zero_delta + pos_delta)/valid_count:.1%})")
    if invalid_count > 0:
        print(f"  Invalid/NA    : {invalid_count} ({invalid_count/total_count:.1%})")

# --- Main Script Logic ---
print("--- Starting Data Loading ---")
train_data_raw = load_jsonl_dataset(os.path.join(DATA_DIR, 'train.jsonl'))
val_data_raw = load_jsonl_dataset(os.path.join(DATA_DIR, 'validation.jsonl'))
test_data_raw = load_jsonl_dataset(os.path.join(DATA_DIR, 'test.jsonl'))
moves_lookup = load_moves_data(MOVES_DIR)
if not train_data_raw or not val_data_raw or not test_data_raw:
    raise ValueError("Dataset loading failed.")

# --- Data Exploration & Preparation ---
print("\n--- Data Exploration & Preparation ---")
train_model_samples, train_no_annotation, train_all = preprocess_data_for_exploration_and_model(train_data_raw)
val_model_samples, _, _ = preprocess_data_for_exploration_and_model(val_data_raw)
test_model_samples, _, _ = preprocess_data_for_exploration_and_model(test_data_raw)

lies = [s for s in train_model_samples if s['label'] == 0]
truths = [s for s in train_model_samples if s['label'] == 1]
print(f"\nUsing {len(train_model_samples)} annotated samples for training/validation.")
num_examples_to_show = 5  # Reduced example print count
print(f"\n--- {num_examples_to_show} Lie Examples (Train Set) ---")
if lies:
    for i, s in enumerate(random.sample(lies, min(num_examples_to_show, len(lies)))):
        print(f"L{i+1} Δ:{s['delta']:<4} | Txt:'{s['text'][:70]}...'")
else:
    print("No lie examples.")
print(f"\n--- {num_examples_to_show} Truth Examples (Train Set) ---")
if truths:
    for i, s in enumerate(random.sample(truths, min(num_examples_to_show, len(truths)))):
        print(f"T{i+1} Δ:{s['delta']:<4} | Txt:'{s['text'][:70]}...'")
else:
    print("No truth examples.")
print(f"\n--- {num_examples_to_show} NOANNOTATION Examples (Train Set) ---")
if train_no_annotation:
    for i, s in enumerate(random.sample(train_no_annotation, min(num_examples_to_show, len(train_no_annotation)))):
        print(f"N{i+1} Sender:{'L' if s.get('label')==0 else 'T' if s.get('label')==1 else '?'} | Txt:'{s['text'][:70]}...'")
else:
    print("No NOANNOTATION examples.")
analyze_delta_scores(lies, "Lies (Train Annotated)")
analyze_delta_scores(truths, "Truths (Train Annotated)")

# --- Final Data Prep for Model ---
train_texts_clean = [s['clean_text'] for s in train_model_samples]
train_features = np.array([extract_action_features(s, moves_lookup) for s in train_model_samples])
train_labels = np.array([s['label'] for s in train_model_samples])
val_texts_clean = [s['clean_text'] for s in val_model_samples]
val_features = np.array([extract_action_features(s, moves_lookup) for s in val_model_samples])
val_labels = np.array([s['label'] for s in val_model_samples])
test_texts_clean = [s['clean_text'] for s in test_model_samples]
test_features = np.array([extract_action_features(s, moves_lookup) for s in test_model_samples])
test_labels = np.array([s['label'] for s in test_model_samples])

if len(train_labels) == 0 or len(val_labels) == 0 or len(test_labels) == 0:
    raise ValueError("Modeling data splits empty.")
print("\nShapes of data prepared for modeling:")
print(f"Train: texts={len(train_texts_clean)}, feats={train_features.shape}, labels={train_labels.shape}")
print(f"Val:   texts={len(val_texts_clean)}, feats={val_features.shape}, labels={val_labels.shape}")
print(f"Test:  texts={len(test_texts_clean)}, feats={test_features.shape}, labels={test_labels.shape}")
num_features = train_features.shape[1]
print(f"Number of features per sample (incl. delta): {num_features}")

# --- Tokenization and Padding ---
print("\nTokenizing and padding text...")
tokenizer = Tokenizer(num_words=MAX_VOCAB_SIZE, oov_token='<OOV>')
tokenizer.fit_on_texts(train_texts_clean)
vocab_size = min(MAX_VOCAB_SIZE, len(tokenizer.word_index) + 1)
print(f"Vocab size:{vocab_size}")
train_sequences = tokenizer.texts_to_sequences(train_texts_clean)
val_sequences = tokenizer.texts_to_sequences(val_texts_clean)
test_sequences = tokenizer.texts_to_sequences(test_texts_clean)
train_padded = pad_sequences(train_sequences, maxlen=MAX_SEQUENCE_LENGTH, padding='post', truncating='post')
val_padded = pad_sequences(val_sequences, maxlen=MAX_SEQUENCE_LENGTH, padding='post', truncating='post')
test_padded = pad_sequences(test_sequences, maxlen=MAX_SEQUENCE_LENGTH, padding='post', truncating='post')
print(f"Padded shapes: Train:{train_padded.shape}, Val:{val_padded.shape}, Test:{test_padded.shape}")

# --- Data Balancing ---
if BALANCE_METHOD == 'downsample':
    print("\nBalancing training data (downsample)...")
    train_df = pd.DataFrame({'pad': list(train_padded), 'feat': list(train_features), 'lbl': train_labels})
    df_maj = train_df[train_df.lbl == 1]
    df_min = train_df[train_df.lbl == 0]
    print(f"Orig counts: Truth(1):{len(df_maj)}, Deceptive(0):{len(df_min)}")
    if len(df_min) == 0:
        print("Warn: Minority empty.")
        bal_train_pad, bal_train_feat, bal_train_lbl = train_padded, train_features, train_labels
    else:
        df_maj_ds = resample(df_maj, replace=False, n_samples=len(df_min), random_state=SEED)
        df_bal = pd.concat([df_min, df_maj_ds]).sample(frac=1, random_state=SEED).reset_index(drop=True)
        bal_train_pad = np.array(df_bal['pad'].tolist())
        bal_train_feat = np.array(df_bal['feat'].tolist())
        bal_train_lbl = np.array(df_bal['lbl'].tolist())
        print(f"Balanced counts: Truth(1):{sum(bal_train_lbl==1)}, Deceptive(0):{sum(bal_train_lbl==0)}")
    print(f"Shapes after balancing: padded={bal_train_pad.shape}, feats={bal_train_feat.shape}, labels={bal_train_lbl.shape}")
else:
    print("\nNo balancing.")
    bal_train_pad, bal_train_feat, bal_train_lbl = train_padded, train_features, train_labels
if len(bal_train_lbl) == 0:
    raise ValueError("Train data empty after balancing.")

# --- Model Definition (Simple Transformer + Features + Delta) ---
def create_simple_transformer_model_with_features(vocab_size, embedding_dim, num_heads, max_sequence_length, num_total_features, dropout_rate=0.1):
    text_input = Input(shape=(max_sequence_length,), name='text_input')
    embedding = Embedding(input_dim=vocab_size, output_dim=embedding_dim)(text_input)
    attn_output = MultiHeadAttention(num_heads=num_heads, key_dim=embedding_dim)(query=embedding, value=embedding, key=embedding)
    attn_output = Dropout(dropout_rate)(attn_output)
    out1 = LayerNormalization(epsilon=1e-6)(Add()([embedding, attn_output]))
    text_branch = GlobalAveragePooling1D()(out1)
    text_branch = Dropout(0.4)(text_branch)
    feature_input = Input(shape=(num_total_features,), name='feature_input')
    concatenated = Concatenate()([text_branch, feature_input])
    dense1 = Dense(64, activation='relu')(concatenated)
    dropout1 = Dropout(0.5)(dense1)
    outputs = Dense(1, activation='sigmoid', name='output')(dropout1)
    model = Model(inputs=[text_input, feature_input], outputs=outputs, name="SimpleTransformer_plus_Features_and_Delta")
    return model

# --- Compile and Train ---
print("\n--- Training Simple Transformer + Action Features + Delta (Downsampled) ---")
model_name = "SimpleTransformer_plus_Features_Delta_Downsampled"
model = create_simple_transformer_model_with_features(
    vocab_size, EMBEDDING_DIM, NUM_HEADS, MAX_SEQUENCE_LENGTH, num_features, DROPOUT_RATE
)
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE), loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

model_filename = model_name.lower() + ".keras"
model_path = os.path.join(OUTPUT_DIR, f"best_{model_filename}")
early_stopping = EarlyStopping(monitor='val_loss', patience=PATIENCE, verbose=1, restore_best_weights=True)
model_checkpoint = ModelCheckpoint(model_path, monitor='val_loss', save_best_only=True, mode='min', verbose=1)
print(f"\nTraining {model_name}...")
history = model.fit([bal_train_pad, bal_train_feat], bal_train_lbl, batch_size=BATCH_SIZE, epochs=EPOCHS,
                    validation_data=([val_padded, val_features], val_labels), callbacks=[early_stopping, model_checkpoint], verbose=1)

# --- Evaluation ---
print(f"\n--- Evaluating Best {model_name} on Test Set ---")
print(f"Loading best weights from {model_path}")
try:
    model.load_weights(model_path)
except Exception as e:
    print(f"Warn: Load weights fail {model_path}: {e}")

test_inputs = [test_padded, test_features]
test_loss, test_accuracy = model.evaluate(test_inputs, test_labels, verbose=0)
test_pred_probs = model.predict(test_inputs)
test_pred_bin = (test_pred_probs > 0.5).astype(int)

from sklearn.metrics import precision_score, recall_score
prec = precision_score(test_labels, test_pred_bin, pos_label=1, zero_division=0)
rec = recall_score(test_labels, test_pred_bin, pos_label=1, zero_division=0)
f1 = f1_score(test_labels, test_pred_bin, pos_label=1, zero_division=0)
f1_macro = f1_score(test_labels, test_pred_bin, average='macro', zero_division=0)
f1_lie = f1_score(test_labels, test_pred_bin, pos_label=0, average='binary', zero_division=0)
results = {model_name: {"Accuracy": test_accuracy, "Macro F1": f1_macro, "Lie F1 (0)": f1_lie, "F1 (1)": f1,
                         "Prec (1)": prec, "Rec (1)": rec, "Loss": test_loss}}
print(f"\n[{model_name}] Test Set Metrics:")
print(f" Accuracy:{test_accuracy:.4f}, MacroF1:{f1_macro:.4f}, LieF1:{f1_lie:.4f}, Loss:{test_loss:.4f}")

conf_matrix = confusion_matrix(test_labels, test_pred_bin)
plt.figure(figsize=(7, 5))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Pred Decep(0)', 'Pred Truth(1)'],
            yticklabels=['Actual Decep(0)', 'Actual Truth(1)'])
plt.title(f'[{model_name}]\nMacro F1:{f1_macro:.4f}|Lie F1:{f1_lie:.4f}', fontsize=10)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plot_path = os.path.join(OUTPUT_DIR, f"cm_{model_filename.replace('.keras','.png')}")
plt.savefig(plot_path)
print(f"Saved CM to {plot_path}")
plt.show()

print("\n--- Final Model Summary (Test Set) ---")
if results:
    results_df = pd.DataFrame(results).T
    cols_order = ["Accuracy", "Macro F1", "Lie F1 (0)", "F1 (1)", "Prec (1)", "Rec (1)", "Loss", "Error"]
    results_df = results_df.reindex(columns=cols_order)
    print(results_df.to_markdown(floatfmt=".4f"))
else:
    print("No models eval.")
print("\n--- Script Finished ---")

# %% [code] Additional Visualization: Bar Charts and Pie Charts for Delta Score Distributions
# Create a DataFrame from training samples with valid labels
df = pd.DataFrame(train_model_samples)
df = df[df['label'] != -1]  # Filter out invalid samples

# Convert delta column to numeric (if not already done) and create absolute delta column
df['delta_numeric'] = pd.to_numeric(df['delta'], errors='coerce')
df['abs_delta'] = df['delta_numeric'].abs()

# Define mapping for label values to names
label_mapping = {1: 'Truth', 0: 'Lie'}

# For each label (Truth and Lie) and for each instance (past, present, future)
for label_val, label_name in label_mapping.items():
    for instance in ['past', 'present', 'future']:
        subset = df[(df['label'] == label_val) & (df['relative_instance'] == instance)]
        
        # ----- Bar Chart: Frequency vs Absolute Delta Score -----
        freq = subset['abs_delta'].value_counts().sort_index()
        plt.figure(figsize=(6, 4))
        sns.barplot(x=freq.index, y=freq.values, palette="viridis")
        plt.title(f"{label_name} - {instance.capitalize()} - Frequency of Abs(Delta)")
        plt.xlabel("Absolute Delta Score")
        plt.ylabel("Frequency")
        plt.tight_layout()
        bar_chart_path = os.path.join(OUTPUT_DIR, f"bar_{label_name.lower()}_{instance}.png")
        plt.savefig(bar_chart_path)
        print(f"Saved bar chart to {bar_chart_path}")
        plt.show()
        
        # ----- Pie Chart: Distribution of Negative, Zero, Positive Delta -----
        counts = {}
        counts['Negative'] = subset[subset['delta_numeric'] < 0].shape[0]
        counts['Zero'] = subset[subset['delta_numeric'] == 0].shape[0]
        counts['Positive'] = subset[subset['delta_numeric'] > 0].shape[0]
        plt.figure(figsize=(5, 5))
        plt.pie(counts.values(), labels=counts.keys(), autopct='%1.1f%%',
                colors=sns.color_palette("viridis", n_colors=3))
        plt.title(f"{label_name} - {instance.capitalize()} - Delta Polarity Distribution")
        plt.tight_layout()
        pie_chart_path = os.path.join(OUTPUT_DIR, f"pie_{label_name.lower()}_{instance}.png")
        plt.savefig(pie_chart_path)
        print(f"Saved pie chart to {pie_chart_path}")
        plt.show()
```

```python
# %%
# Import necessary libraries
import pandas as pd
import numpy as np
import json
import os
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
import re
import string
import glob # For finding files
import warnings
import random

# Scikit-learn imports
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix
from sklearn.utils import resample

# TensorFlow / Keras imports
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import (
    Input, Embedding, Dense, Dropout, Flatten, Concatenate,
    Layer, MultiHeadAttention, LayerNormalization, GlobalAveragePooling1D, Add
)
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras import backend as K

# Filter out specific warnings if they are noisy and understood
warnings.filterwarnings("ignore", category=UserWarning, module='tensorflow')
warnings.filterwarnings("ignore", category=RuntimeWarning) # Ignore potential runtime warnings in np operations

# Set random seeds for reproducibility
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

# --- Constants and Configuration ---
BASE_DATA_DIR = '/kaggle/input/diplomacy/transformers/default/1/Dataset/' # Or your local path
if not os.path.exists(BASE_DATA_DIR) and 'kaggle' not in BASE_DATA_DIR:
    print("Warning: BASE_DATA_DIR not found, attempting relative path './Dataset/'")
    BASE_DATA_DIR = './Dataset/'
    if not os.path.exists(BASE_DATA_DIR):
         raise FileNotFoundError("Cannot find Dataset directory. Please check BASE_DATA_DIR path.")

DATA_DIR = os.path.join(BASE_DATA_DIR, 'data')
MOVES_DIR = os.path.join(BASE_DATA_DIR, 'moves')
OUTPUT_DIR = '/kaggle/working/'
if not os.path.exists(OUTPUT_DIR):
     OUTPUT_DIR = './output/'
     print(f"Warning: Kaggle directory not found, using local directory: {OUTPUT_DIR}")

# Model Hyperparameters
MAX_VOCAB_SIZE = 10000
MAX_SEQUENCE_LENGTH = 60
EMBEDDING_DIM = 128
NUM_HEADS = 4
DROPOUT_RATE = 0.1
# Training Params
BATCH_SIZE = 32
EPOCHS = 25 # Increased epochs slightly, maybe beneficial with new feature
PATIENCE = 5 # Keep patience the same
LEARNING_RATE = 1e-4
BALANCE_METHOD = 'downsample'

os.makedirs(OUTPUT_DIR, exist_ok=True)

# --- Territory/Keyword Definitions ---
# (Not used in feature extraction for now)
# ... (keep the definitions if you might use them later)

# --- Helper Functions ---

def clean_text(text):
    # (Same as before)
    if not isinstance(text, str): return ""
    text = text.lower()
    text = re.sub(r'http\S+', '', text); text = re.sub(r'@\w+', '', text); text = re.sub(r'#\w+', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation.replace("'", ""))); text = re.sub(r'\d+', '', text); text = re.sub(r'\s+', ' ', text).strip()
    return text

def load_jsonl_dataset(file_path):
    # (Same as before)
    data = []; skipped_lines = 0
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            for i, line in enumerate(f):
                try: data.append(json.loads(line))
                except json.JSONDecodeError: skipped_lines += 1
    except FileNotFoundError: print(f"Error: File not found at {file_path}"); return []
    if skipped_lines > 0: print(f"Warning: Skipped {skipped_lines} invalid JSON lines in {file_path}")
    return data

def load_moves_data(moves_dir):
    # (Same as before, technically not needed for current features but good practice)
    moves_lookup = {}; print(f"Loading moves data from: {moves_dir}")
    if not os.path.isdir(moves_dir): print(f"Error: Moves directory not found: {moves_dir}"); return {}
    loaded_count=0; file_count=0; error_files=0
    for filepath in glob.glob(os.path.join(moves_dir, '*.json')):
        file_count+=1; filename=os.path.basename(filepath); match=re.match(r'DiplomacyGame(\d+)_(\d+)_(\w+)\.json', filename)
        if match:
            try: game_id=int(match.group(1)); year=match.group(2); season=match.group(3).lower();
            except ValueError: error_files+=1; continue
            try:
                with open(filepath,'r',encoding='utf-8') as f: moves_lookup[(game_id,year,season)]=json.load(f); loaded_count+=1
            except Exception as e: error_files+=1
        else: error_files += 1
    print(f"Successfully loaded move data for {loaded_count} game turns from {file_count} files found.")
    if error_files > 0: print(f"  Encountered issues with {error_files} files.")
    return moves_lookup

# --- UPDATED Preprocessing Function ---
def preprocess_data_for_exploration_and_model(data_raw):
    all_processed_samples = []
    annotated_samples_for_model = []
    no_annotation_samples = []
    skipped_dialogues = 0
    total_messages_processed = 0

    for dialogue_idx, dialogue_sample in enumerate(data_raw):
        messages = dialogue_sample.get('messages', [])
        sender_labels = dialogue_sample.get('sender_labels', [])
        receiver_labels = dialogue_sample.get('receiver_labels', [])
        speakers = dialogue_sample.get('speakers', [])
        recipients = dialogue_sample.get('receivers', [])
        game_id = dialogue_sample.get('game_id')
        years = dialogue_sample.get('years', [])
        seasons = dialogue_sample.get('seasons', [])
        game_scores = dialogue_sample.get('game_score', [])
        game_score_deltas = dialogue_sample.get('game_score_delta', [])

        list_len = len(messages)
        total_messages_processed += list_len

        required_lists = [sender_labels, receiver_labels, speakers, recipients, years, seasons, game_scores, game_score_deltas]
        if not game_id or not list_len > 0 or not all(len(lst) == list_len for lst in required_lists):
            skipped_dialogues += 1
            continue

        for i in range(list_len):
            # --- Calculate current delta ---
            delta_numeric = 0.0
            delta_str = game_score_deltas[i] if game_score_deltas and i < len(game_score_deltas) else 'N/A'
            if isinstance(delta_str, (int, float)): delta_numeric = float(delta_str)
            elif isinstance(delta_str, str):
                try: delta_numeric = float(delta_str)
                except (ValueError, TypeError): delta_str = 'N/A'

            # --- Store numeric score for speaker ---
            score_numeric = np.nan # Use NaN for missing/invalid scores initially
            score_str = game_scores[i] if game_scores and i < len(game_scores) else 'N/A'
            if isinstance(score_str, (int, float)): score_numeric = float(score_str)
            elif isinstance(score_str, str):
                try: score_numeric = float(score_str)
                except (ValueError, TypeError): score_str = 'N/A'

            # Determine binary label
            label = -1
            if sender_labels[i] is False: label = 0
            elif sender_labels[i] is True: label = 1

            sample = {
                'text': messages[i] or "", 'clean_text': clean_text(messages[i] or ""),
                'game_id': game_id, 'speaker': str(speakers[i] or 'unknown').lower(),
                'recipient': str(recipients[i] or 'unknown').lower(), 'year': str(years[i] or 'unknown'),
                'season': str(seasons[i] or 'unknown').lower(),
                'score': score_str, 'score_numeric': score_numeric, # Store both raw and numeric score
                'delta': delta_str, 'delta_numeric': delta_numeric, # Store both raw and numeric delta
                'receiver_annotation': receiver_labels[i], 'sender_label_raw': sender_labels[i],
                'label': label
            }

            all_processed_samples.append(sample)
            if sample['receiver_annotation'] == "NOANNOTATION":
                no_annotation_samples.append(sample)
            if sample['label'] != -1 :
                annotated_samples_for_model.append(sample)

    print(f"Preprocessed {len(all_processed_samples)} total messages from {dialogue_idx+1-skipped_dialogues} valid dialogues.")
    if skipped_dialogues > 0: print(f"  Warning: Skipped {skipped_dialogues} dialogues due to inconsistent data.")
    print(f"  Found {len(no_annotation_samples)} 'NOANNOTATION' receiver labels.")
    print(f"  Found {len(annotated_samples_for_model)} samples with valid annotations for modeling.")

    return annotated_samples_for_model, no_annotation_samples, all_processed_samples

# --- Helper function to create the Score Lookup ---
def create_score_lookup(all_samples):
    """ Creates a lookup dictionary for player scores. """
    score_lookup = defaultdict(lambda: np.nan) # Default to NaN if score not found
    processed_keys = set()
    print(f"Creating score lookup from {len(all_samples)} processed samples...")
    skipped_no_score = 0
    valid_scores = 0
    for sample in all_samples:
        game_id = sample.get('game_id')
        year = sample.get('year')
        season = sample.get('season')
        player = sample.get('speaker') # Use speaker's score for the lookup
        score_num = sample.get('score_numeric')

        # Ensure key components are valid and score is numeric
        if game_id and year != 'unknown' and season != 'unknown' and player != 'unknown':
            key = (game_id, year, season, player)
             # Store first valid score found for a player in a turn
            if key not in processed_keys:
                 if pd.notna(score_num): # Check if score_num is not NaN
                     score_lookup[key] = score_num
                     processed_keys.add(key)
                     valid_scores+=1
                 else:
                    skipped_no_score +=1

    print(f"Score lookup created with {len(score_lookup)} valid entries ({valid_scores} unique player-turn scores).")
    if skipped_no_score>0: print(f"  Skipped {skipped_no_score} instances due to missing/invalid numeric scores.")
    return score_lookup

# --- Helper function to get the next turn ---
def get_next_turn(current_year_str, current_season_str):
    """Calculates the next game turn (year, season)."""
    try:
        current_year = int(current_year_str)
        current_season = current_season_str.lower()

        if current_season == 'spring':
            return str(current_year), 'fall'
        elif current_season == 'fall':
            return str(current_year), 'winter'
        elif current_season == 'winter':
            return str(current_year + 1), 'spring'
        else:
            # Handle unknown/invalid season - return None or raise error? Returning None seems safer.
            return None, None
    except (ValueError, TypeError):
         # Handle non-integer year or non-string season
        return None, None

# --- MODIFIED Feature Extraction function ---
def extract_action_features(sample_info, score_lookup):
    """
    Extracts features:
    1. Current numeric game score delta.
    2. Calculated future game score delta between the same two players.

    Args:
        sample_info (dict): Dictionary containing message details.
        score_lookup (dict): Pre-calculated lookup for player scores by turn.
                               Key: (game_id, year, season, player), Value: numeric_score

    Returns:
        list: A list containing two floats: [current_delta, future_delta].
              Defaults to 0.0 for missing values.
    """
    # 1. Get current delta (default to 0.0 if missing)
    current_delta = sample_info.get('delta_numeric', 0.0)
    if not isinstance(current_delta, (int, float)) or pd.isna(current_delta):
        current_delta = 0.0 # Ensure it's a valid number

    # 2. Calculate future delta
    future_delta = 0.0 # Default future delta

    game_id = sample_info.get('game_id')
    current_year = sample_info.get('year')
    current_season = sample_info.get('season')
    speaker = sample_info.get('speaker')
    recipient = sample_info.get('recipient')

    if game_id and current_year != 'unknown' and current_season != 'unknown' and speaker != 'unknown' and recipient != 'unknown':
        next_year, next_season = get_next_turn(current_year, current_season)

        if next_year and next_season:
            # Lookup scores for speaker and recipient in the NEXT turn
            future_speaker_score = score_lookup.get((game_id, next_year, next_season, speaker), np.nan)
            future_recipient_score = score_lookup.get((game_id, next_year, next_season, recipient), np.nan)

            # Check if both scores were found and are numeric before calculating delta
            if pd.notna(future_speaker_score) and pd.notna(future_recipient_score):
                try:
                    # Ensure they are floats before subtracting
                    future_delta = float(future_speaker_score) - float(future_recipient_score)
                except (ValueError, TypeError):
                    future_delta = 0.0 # Default if conversion fails unexpectedly
            # else: Keep future_delta as default 0.0 if scores are missing

    return [current_delta, future_delta]


# Function for Delta Score Analysis (no changes needed)
def analyze_delta_scores(samples_list, label_name, delta_key='delta_numeric'):
    # Analyzes the distribution of numeric delta scores for a given list of samples.
    deltas_numeric = []
    invalid_count = 0
    for sample in samples_list:
        delta_num = sample.get(delta_key) # Use the specified key
        if isinstance(delta_num, (int, float)) and pd.notna(delta_num): # Check if it's actually numeric & not NaN
             deltas_numeric.append(delta_num)
        else: # Counts samples where delta_numeric wasn't successfully parsed earlier or is NaN
             invalid_count+=1
             # print(f"Debug: Invalid {delta_key} found: {delta_num} in sample for {label_name}")


    total_count = len(samples_list)
    valid_count = len(deltas_numeric)

    if total_count == 0:
        print(f"\nNo samples provided for Delta Score Analysis for '{label_name}' using '{delta_key}'.")
        return

    # Calculate counts for different delta ranges
    neg_delta = sum(1 for d in deltas_numeric if d < 0)
    zero_delta = sum(1 for d in deltas_numeric if d == 0)
    pos_delta = sum(1 for d in deltas_numeric if d > 0)

    print(f"\n--- {delta_key.replace('_',' ').title()} Analysis for '{label_name}' (Total Samples: {total_count}) ---")
    print(f"  Valid Numeric Values: {valid_count} ({valid_count/total_count:.1%})")
    if invalid_count > 0:
      print(f"  Invalid/Non-Numeric: {invalid_count} ({invalid_count/total_count:.1%})")

    if valid_count > 0:
      print(f"  Distribution of Valid Values:")
      print(f"    Negative (< 0): {neg_delta:>5} ({neg_delta/valid_count:.1%})")
      print(f"    Zero     (= 0): {zero_delta:>5} ({zero_delta/valid_count:.1%})")
      print(f"    Positive (> 0): {pos_delta:>5} ({pos_delta/valid_count:.1%})")
      print(f"    ----------")
      print(f"    Non-Neg (>= 0): {zero_delta + pos_delta:>5} ({(zero_delta + pos_delta)/valid_count:.1%})")
      print(f"    Non-Pos (<= 0): {zero_delta + neg_delta:>5} ({(zero_delta + neg_delta)/valid_count:.1%})")
    else:
      print("  No valid numeric values found to analyze distribution.")

# --- Main Script Logic ---

print("--- Starting Data Loading ---")
train_data_raw = load_jsonl_dataset(os.path.join(DATA_DIR, 'train.jsonl'))
val_data_raw = load_jsonl_dataset(os.path.join(DATA_DIR, 'validation.jsonl'))
test_data_raw = load_jsonl_dataset(os.path.join(DATA_DIR, 'test.jsonl'))
moves_lookup = load_moves_data(MOVES_DIR) # Load moves (though not used for features now)
if not train_data_raw or not val_data_raw or not test_data_raw: raise ValueError("Dataset loading failed.")

print("\n--- Data Preprocessing & Exploration ---")
train_model_samples, train_no_annotation, train_all = preprocess_data_for_exploration_and_model(train_data_raw)
val_model_samples, _, val_all = preprocess_data_for_exploration_and_model(val_data_raw)
test_model_samples, _, test_all = preprocess_data_for_exploration_and_model(test_data_raw)
if not train_model_samples: raise ValueError("Critical error: No valid samples for training after preprocessing.")

# --- Create Score Lookup from ALL processed samples ---
# We use train_all+val_all+test_all to get the most comprehensive score history possible
all_samples_combined = train_all + val_all + test_all
score_lookup = create_score_lookup(all_samples_combined)

# --- Prepare Data for Model Input (including feature extraction) ---
print("\n--- Preparing Data for Modeling (Extracting Current & Future Delta) ---")

# Helper function to apply feature extraction and store results
def prepare_model_input(samples, score_lookup_dict):
    texts = [s['clean_text'] for s in samples]
    # Pass the created score_lookup to the feature extractor
    features_list = [extract_action_features(s, score_lookup_dict) for s in samples]
    features = np.array(features_list).astype(np.float32) # Ensure float32 for TF
    labels = np.array([s['label'] for s in samples])

    # Store future delta separately for analysis before balancing
    future_deltas_analysis = [f[1] for f in features_list] # Get the second element (future delta)

    return texts, features, labels, future_deltas_analysis

# Prepare train, validation, and test sets
train_texts_clean, train_features, train_labels, train_future_deltas = prepare_model_input(train_model_samples, score_lookup)
val_texts_clean, val_features, val_labels, _ = prepare_model_input(val_model_samples, score_lookup) # No need for val future delta analysis
test_texts_clean, test_features, test_labels, _ = prepare_model_input(test_model_samples, score_lookup) # No need for test future delta analysis

# --- Analyze Future Delta Distribution (on Train set before balancing) ---
# Add future delta to the samples dict temporarily for analysis function
for i, sample in enumerate(train_model_samples):
    sample['future_delta_numeric'] = train_future_deltas[i]

lies_train = [s for s in train_model_samples if s['label'] == 0]
truths_train = [s for s in train_model_samples if s['label'] == 1]
analyze_delta_scores(lies_train, "Lies (Train Set, Label 0)", delta_key='future_delta_numeric')
analyze_delta_scores(truths_train, "Truths (Train Set, Label 1)", delta_key='future_delta_numeric')


print("\nShapes of data prepared for modeling:")
print(f"Train: texts={len(train_texts_clean)}, features={train_features.shape}, labels={train_labels.shape}")
print(f"Val:   texts={len(val_texts_clean)}, features={val_features.shape}, labels={val_labels.shape}")
print(f"Test:  texts={len(test_texts_clean)}, features={test_features.shape}, labels={test_labels.shape}")

if train_features.ndim == 2:
    num_features = train_features.shape[1]
    print(f"Number of features per sample: {num_features}")
    if num_features != 2:
        print(f"WARNING: Expected 2 features (current+future delta), but found {num_features}.")
else:
    raise ValueError(f"Features array has unexpected shape: {train_features.shape}.")


# --- Text Tokenization and Padding ---
print("\nTokenizing and padding text sequences...")
tokenizer = Tokenizer(num_words=MAX_VOCAB_SIZE, oov_token='<OOV>')
tokenizer.fit_on_texts(train_texts_clean)
word_index_len = len(tokenizer.word_index)
vocab_size = min(MAX_VOCAB_SIZE, word_index_len + 1)
print(f"Tokenizer fitted on {word_index_len} words. Effective vocab size: {vocab_size}")

train_sequences=tokenizer.texts_to_sequences(train_texts_clean); val_sequences=tokenizer.texts_to_sequences(val_texts_clean); test_sequences=tokenizer.texts_to_sequences(test_texts_clean)
train_padded=pad_sequences(train_sequences,maxlen=MAX_SEQUENCE_LENGTH,padding='post',truncating='post')
val_padded=pad_sequences(val_sequences,maxlen=MAX_SEQUENCE_LENGTH,padding='post',truncating='post'); test_padded=pad_sequences(test_sequences,maxlen=MAX_SEQUENCE_LENGTH,padding='post',truncating='post')
print(f"Padded sequence shapes: Train={train_padded.shape}, Val={val_padded.shape}, Test={test_padded.shape}")


# --- Data Balancing ---
print(f"\nApplying data balancing strategy: '{BALANCE_METHOD}'")
if BALANCE_METHOD == 'downsample':
    print("Balancing training data (downsample)...")
    train_df = pd.DataFrame({'pad': list(train_padded), 'feat': list(train_features), 'lbl': train_labels})
    df_maj=train_df[train_df.lbl==1]; df_min=train_df[train_df.lbl==0]
    print(f"Original counts: Truth(1):{len(df_maj)}, Lie(0):{len(df_min)}")
    if len(df_min)==0 or len(df_maj)==0:
        print("Warn: Minority or Majority class empty. Skip downsample.")
        bal_train_pad, bal_train_feat, bal_train_lbl = train_padded, train_features, train_labels
    else:
        df_maj_ds=resample(df_maj,replace=False,n_samples=len(df_min),random_state=SEED)
        df_bal=pd.concat([df_min,df_maj_ds]).sample(frac=1,random_state=SEED).reset_index(drop=True)
        bal_train_pad=np.array(df_bal['pad'].tolist()); bal_train_feat=np.array(df_bal['feat'].tolist()); bal_train_lbl=np.array(df_bal['lbl'].tolist())
        print(f"Balanced counts: Truth(1):{sum(bal_train_lbl==1)}, Lie(0):{sum(bal_train_lbl==0)}")
    print(f"Shapes after balancing: padded={bal_train_pad.shape}, feats={bal_train_feat.shape}, labels={bal_train_lbl.shape}")
else:
    print("\nNo balancing.")
    bal_train_pad,bal_train_feat,bal_train_lbl = train_padded,train_features,train_labels

if len(bal_train_lbl)==0: raise ValueError("Train data empty after balancing.")


# --- Model Definition (Simple Transformer + Current & Future Delta) ---
# Updated function name
def create_transformer_model_with_delta_features(
    vocab_size, embedding_dim, num_heads, max_sequence_length, num_total_features, dropout_rate=0.1
    ):
    """ Creates Keras model combining text embeddings with TWO numeric features (current+future delta). """
    if num_total_features != 2:
         print(f"Model Definition Warning: Expected num_total_features=2, but got {num_total_features}")

    text_input = Input(shape=(max_sequence_length,), name='text_input')
    embedding_layer = Embedding(input_dim=vocab_size, output_dim=embedding_dim, name='embedding')(text_input)
    attention_output = MultiHeadAttention(num_heads=num_heads, key_dim=embedding_dim, name='attention')(
        query=embedding_layer, value=embedding_layer, key=embedding_layer)
    attention_output = Dropout(dropout_rate, name='attn_dropout')(attention_output)
    add_norm_1 = LayerNormalization(epsilon=1e-6, name='add_norm_1')(Add(name='add_resid_1')([embedding_layer, attention_output]))
    text_vector = GlobalAveragePooling1D(name='global_avg_pooling')(add_norm_1)
    text_vector = Dropout(0.4, name='text_dropout')(text_vector)

    # Feature Input Branch (expects shape (2,))
    feature_input = Input(shape=(num_total_features,), name='feature_input')
    # Optional: Normalize features if needed: features_norm = LayerNormalization()(feature_input)

    # Concatenate text vector and feature vector(s)
    concatenated = Concatenate(name='concatenate')([text_vector, feature_input]) # Use feature_input directly

    dense_1 = Dense(64, activation='relu', name='dense_1')(concatenated)
    dropout_1 = Dropout(0.5, name='dense_dropout')(dense_1)
    output = Dense(1, activation='sigmoid', name='output')(dropout_1)

    model = Model(inputs=[text_input, feature_input], outputs=output, name="Transformer_CurrentFutureDelta")
    return model

# --- Compile and Train ---
print("\n--- Defining and Compiling Model (Transformer + Current & Future Delta) ---")
model_name = "Transformer_CurrentFutureDelta_Downsampled" # Updated name

# Create the model (num_features should be 2 now)
model = create_transformer_model_with_delta_features(
    vocab_size=vocab_size, embedding_dim=EMBEDDING_DIM, num_heads=NUM_HEADS,
    max_sequence_length=MAX_SEQUENCE_LENGTH, num_total_features=num_features, # Should be 2
    dropout_rate=DROPOUT_RATE
)

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE),
              loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# Setup Callbacks
model_filename = model_name.lower() + ".keras"
model_path = os.path.join(OUTPUT_DIR, f"best_{model_filename}")
early_stopping = EarlyStopping(monitor='val_loss',patience=PATIENCE,verbose=1,restore_best_weights=True)
model_checkpoint = ModelCheckpoint(model_path,monitor='val_loss',save_best_only=True,mode='min',verbose=1)

print(f"\n--- Starting Training: {model_name} ---")
print(f"Model checkpoints will be saved to: {model_path}")
history = model.fit(
    [bal_train_pad, bal_train_feat], bal_train_lbl,
    batch_size=BATCH_SIZE, epochs=EPOCHS,
    validation_data=([val_padded, val_features], val_labels),
    callbacks=[early_stopping, model_checkpoint], verbose=1
)

# --- Evaluation ---
print(f"\n--- Evaluating Best '{model_name}' on Test Set ---")
print(f"Loading best model weights from: {model_path}")
try:
    if os.path.exists(model_path): model.load_weights(model_path); print("Best weights loaded.")
    else: print("Warn: Model file not found. Using end-of-training weights.")
except Exception as e: print(f"Warn: Failed loading weights: {e}")

test_inputs = [test_padded, test_features]
test_loss, test_accuracy = model.evaluate(test_inputs, test_labels, verbose=0)
test_pred_probs = model.predict(test_inputs)
test_pred_binary = (test_pred_probs > 0.5).astype(int).flatten()

# Calculate metrics (same as before)
precision_truth = precision_score(test_labels, test_pred_binary, pos_label=1, zero_division=0)
recall_truth = recall_score(test_labels, test_pred_binary, pos_label=1, zero_division=0)
f1_truth = f1_score(test_labels, test_pred_binary, pos_label=1, zero_division=0)
precision_lie = precision_score(test_labels, test_pred_binary, pos_label=0, zero_division=0)
recall_lie = recall_score(test_labels, test_pred_binary, pos_label=0, zero_division=0)
f1_lie = f1_score(test_labels, test_pred_binary, pos_label=0, zero_division=0)
f1_macro = f1_score(test_labels, test_pred_binary, average='macro', zero_division=0)
f1_weighted = f1_score(test_labels, test_pred_binary, average='weighted', zero_division=0)

results = {
    model_name: {
        "Accuracy": test_accuracy, "Macro F1": f1_macro, "Weighted F1": f1_weighted,
        "Lie F1 (0)": f1_lie, "Truth F1 (1)": f1_truth, "Lie Recall (0)": recall_lie,
        "Truth Recall (1)": recall_truth, "Lie Precision (0)": precision_lie,
        "Truth Precision (1)": precision_truth, "Loss": test_loss }}

print(f"\n[{model_name}] Test Set Metrics:")
print(f"  Accuracy:    {test_accuracy:.4f}"); print(f"  Macro F1:    {f1_macro:.4f}"); print(f"  Weighted F1: {f1_weighted:.4f}")
print(f"  Lie F1 (0):  {f1_lie:.4f} (Recall: {recall_lie:.4f}, Precision: {precision_lie:.4f})")
print(f"  Truth F1 (1):{f1_truth:.4f} (Recall: {recall_truth:.4f}, Precision: {precision_truth:.4f})"); print(f"  Loss:        {test_loss:.4f}")

# Confusion Matrix Plotting
conf_matrix = confusion_matrix(test_labels, test_pred_binary)
plt.figure(figsize=(6, 5));
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Predicted Lie (0)', 'Predicted Truth (1)'], yticklabels=['Actual Lie (0)', 'Actual Truth (1)'])
plt.title(f'CM: {model_name}\nAcc:{test_accuracy:.4f}|Macro F1:{f1_macro:.4f}|Lie F1:{f1_lie:.4f}', fontsize=10)
plt.ylabel('Actual Label'); plt.xlabel('Predicted Label'); plt.tight_layout()
plot_filename = f"cm_{model_filename.replace('.keras','.png')}"
plot_path = os.path.join(OUTPUT_DIR, plot_filename)
try: plt.savefig(plot_path); print(f"Saved CM plot to: {plot_path}")
except Exception as e: print(f"Error saving CM plot: {e}")
plt.show()

# Final Summary Table
print("\n--- Final Model Evaluation Summary (Test Set) ---")
if results:
    results_df = pd.DataFrame(results).T
    cols_order = ["Accuracy", "Macro F1", "Lie F1 (0)", "Truth F1 (1)", "Lie Recall (0)", "Truth Recall (1)", "Lie Precision (0)", "Truth Precision (1)", "Weighted F1", "Loss"]
    results_df = results_df.reindex(columns=cols_order)
    print(results_df.to_markdown(floatfmt=".4f"))
else: print("No models evaluated.")

print("\n--- Script Finished ---")
```

```python
# %%
# --- Imports ---
import pandas as pd
import numpy as np
import json
import os
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
import re
import string
import glob
import warnings
import random

# Scikit-learn imports
from sklearn.metrics import f1_score, confusion_matrix, classification_report
from sklearn.utils import resample # Added for downsampling

# TensorFlow / Keras imports
import tensorflow as tf
from tensorflow.keras.layers import (Input, Embedding, Dense, Dropout, Concatenate,
                                     Layer, MultiHeadAttention, LayerNormalization,
                                     GlobalAveragePooling1D, Add, Reshape,
                                     Multiply)
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# --- Constants and Configuration ---
SEED = 42
random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)
warnings.filterwarnings("ignore", category=RuntimeWarning)
warnings.filterwarnings("ignore", category=UserWarning) # Less verbose about build

MAX_PLAYERS = 7; MAX_SEQUENCE_LENGTH = 60; EMBEDDING_DIM = 128; MAX_VOCAB_SIZE = 10000
NUM_HEADS_TXT=4; NUM_HEADS_REASON=4; DROPOUT_RATE=0.1; GCN_UNITS=64
DELTA_FEATURE_DIM=2
LEARNING_RATE=1e-4; BATCH_SIZE=32; EPOCHS=30;
PATIENCE=8

BASE_DATA_DIR = '/kaggle/input/diplomacy/transformers/default/1/Dataset/'
if not os.path.exists(BASE_DATA_DIR) and 'kaggle' not in BASE_DATA_DIR: BASE_DATA_DIR = './Dataset/'
DATA_DIR = os.path.join(BASE_DATA_DIR, 'data')
OUTPUT_DIR = '/kaggle/working/' if os.path.exists('/kaggle/working/') else './output/'
os.makedirs(OUTPUT_DIR, exist_ok=True)

# --- Helper Functions ---

def load_jsonl_dataset(file_path):
    """Loads data from a JSON Lines file, handling errors."""
    data = []
    skipped_lines = 0
    try: # Correct indentation for try
        with open(file_path, 'r', encoding='utf-8') as f:
            for i, line in enumerate(f):
                try:
                    data.append(json.loads(line))
                except json.JSONDecodeError:
                    skipped_lines += 1
                    # print(f"Warn: Skipped invalid JSON line {i+1} in {file_path}") # Uncomment for debug
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
        return None # Return None if file not found
    if skipped_lines > 0:
        print(f"Warn: Skipped {skipped_lines} invalid JSON lines in {file_path}")
    return data

def clean_text(text):
    """Cleans text data."""
    if not isinstance(text, str): return ""
    text = text.lower()
    text = re.sub(r'http\S+|@\w+|#\w+', '', text) # Remove urls, mentions, hashtags
    text = text.translate(str.maketrans('', '', string.punctuation.replace("'", ""))) # Remove punctuation except apostrophe
    text = re.sub(r'\d+', '', text) # Remove digits
    text = re.sub(r'\s+', ' ', text).strip() # Remove extra whitespace
    return text

def preprocess_for_hybrid(data_raw):
    """Preprocesses raw data for hybrid model inputs."""
    processed_samples = []
    if data_raw is None: return processed_samples
    skipped_dialogues=0
    for dialogue_idx, dialogue_sample in enumerate(data_raw):
        # Extract necessary fields safely
        game_id=dialogue_sample.get('game_id'); messages=dialogue_sample.get('messages',[]); speakers=dialogue_sample.get('speakers',[]); recipients=dialogue_sample.get('receivers',[])
        years=dialogue_sample.get('years',[]); seasons=dialogue_sample.get('seasons',[]); sender_labels=dialogue_sample.get('sender_labels',[]); game_scores=dialogue_sample.get('game_score',[]); game_score_deltas=dialogue_sample.get('game_score_delta',[])
        list_len=len(messages)
        # Check for essential fields and consistent lengths
        required_lists = [speakers, recipients, years, seasons, sender_labels, game_scores, game_score_deltas]
        # Ensure all elements in required_lists are actually lists before checking length
        if not (game_id and list_len > 0 and all(isinstance(lst, list) and len(lst) == list_len for lst in required_lists if lst is not None)):
             skipped_dialogues+=1
             continue


        for i in range(list_len):
            label=-1; score_num=np.nan; delta_num=np.nan
            # Ensure sender_labels is a list and index i is valid
            if isinstance(sender_labels, list) and i < len(sender_labels):
                if sender_labels[i] is False: label=0
                elif sender_labels[i] is True: label=1

            # Safely access and convert score and delta
            try:
                if game_scores and i < len(game_scores): score_num=float(game_scores[i])
            except (ValueError, TypeError): pass
            try:
                if game_score_deltas and i < len(game_score_deltas): delta_num=float(game_score_deltas[i])
            except (ValueError, TypeError): pass

            # Only keep samples with a valid label (0 or 1)
            if label != -1:
                processed_samples.append({
                    'text': messages[i] if messages and i < len(messages) else "",
                    'clean_text': clean_text(messages[i] if messages and i < len(messages) else ""),
                    'label': label,
                    'game_id': game_id,
                    'year': str(years[i] if years and i < len(years) else 'unknown'),
                    'season': str(seasons[i] if seasons and i < len(seasons) else 'unknown').lower(),
                    'speaker': str(speakers[i] if speakers and i < len(speakers) else 'unknown').lower(),
                    'recipient': str(recipients[i] if recipients and i < len(recipients) else 'unknown').lower(), # Needed for future delta calc
                    'score_numeric': score_num,
                    'delta_numeric': delta_num
                    })
    print(f" Preprocessed {len(processed_samples)} labeled messages.", end="")
    if skipped_dialogues > 0: print(f" (Skipped {skipped_dialogues} dialogues)", end="")
    print()
    return processed_samples

def create_score_lookup(all_samples):
    """Creates a lookup dictionary for player scores."""
    score_lookup=defaultdict(lambda:np.nan); processed_keys=set(); valid_scores=0; skipped_no_score=0
    print(f"Creating score lookup from {len(all_samples)} samples...")
    for sample in all_samples:
        game_id=sample.get('game_id'); year=sample.get('year'); season=sample.get('season'); player=sample.get('speaker'); score_num=sample.get('score_numeric')
        if game_id is not None and year!='unknown' and season!='unknown' and player!='unknown': # Explicit check for game_id None
            key=(game_id,year,season,player)
            if key not in processed_keys:
                if pd.notna(score_num): # Check if score_num is not NaN
                    score_lookup[key]=score_num
                    processed_keys.add(key)
                    valid_scores+=1
                else:
                    skipped_no_score+=1
    print(f"Score lookup created: {len(score_lookup)} entries ({valid_scores} unique). Skipped N/A: {skipped_no_score}.")
    return score_lookup

def get_next_turn(current_year_str, current_season_str): # CORRECTED version
    """Calculates the next game turn (year, season)."""
    try:
        current_year = int(current_year_str)
        current_season = str(current_season_str).lower() # Ensure season is lowercase string
        if current_season == 'spring': return str(current_year), 'fall'
        elif current_season == 'fall': return str(current_year), 'winter'
        elif current_season == 'winter': return str(current_year + 1), 'spring'
        else: return None, None
    except (ValueError, TypeError):
        return None, None

def extract_delta_features(sample_info, score_lookup): # CORRECTED version
    """Extracts current and future delta features for a message."""
    current_delta = 0.0; future_delta = 0.0
    current_delta_val = sample_info.get('delta_numeric', 0.0)
    if not pd.isna(current_delta_val):
         try: current_delta = float(current_delta_val)
         except (ValueError, TypeError): current_delta = 0.0
    game_id=sample_info.get('game_id'); year=sample_info.get('year'); season=sample_info.get('season'); speaker=sample_info.get('speaker'); recipient=sample_info.get('recipient')
    if game_id is not None and year!='unknown' and season!='unknown' and speaker!='unknown' and recipient!='unknown':
        next_year, next_season = get_next_turn(year, season)
        if next_year and next_season:
            fut_spk_score = score_lookup.get((game_id, next_year, next_season, speaker), np.nan)
            fut_rec_score = score_lookup.get((game_id, next_year, next_season, recipient), np.nan)
            if pd.notna(fut_spk_score) and pd.notna(fut_rec_score):
                try: # Correctly indented try-except
                    future_delta = float(fut_spk_score) - float(fut_rec_score)
                except (ValueError, TypeError):
                    future_delta = 0.0 # Fallback
    return np.array([current_delta, future_delta], dtype=np.float32)

def aggregate_data_by_turn(all_processed_samples):
    """Aggregates data by game turn (needed for graph building)."""
    games_data = defaultdict(lambda: defaultdict(lambda: {'messages': [], 'players': set()}))
    for sample in all_processed_samples:
        game_id=sample.get('game_id'); year=sample.get('year'); season=sample.get('season'); speaker=sample.get('speaker'); recipient=sample.get('recipient')
        if game_id is None or year=='unknown' or season=='unknown': continue # Skip if key components missing
        turn_key=(year,season);
        # Only add valid speakers/recipients to players set
        player_set = set()
        if speaker is not None and speaker != 'unknown': player_set.add(speaker)
        if recipient is not None and recipient != 'unknown': player_set.add(recipient)
        games_data[game_id][turn_key]['players'].update(player_set)
        # Append message details for adjacency calc (make sure speaker/recipient aren't None if used later)
        games_data[game_id][turn_key]['messages'].append({'speaker':speaker,'recipient':recipient})

    print(f"Aggregated samples into turns for {len(games_data)} games.")
    return games_data

def build_turn_graph_lookup(games_data_agg, score_lookup):
    """Builds a lookup for turn-specific graph data (X, A+I, M)."""
    print("Building turn graph lookup (using A+I)..."); graph_lookup={}; player_maps_game={}
    default_graph = {'X': np.zeros((MAX_PLAYERS, 1), dtype=np.float32),
                     'A': np.identity(MAX_PLAYERS, dtype=np.float32),
                     'M': np.zeros(MAX_PLAYERS, dtype=bool)}
    for game_id, turns in games_data_agg.items():
        # Calculate player map for this game
        all_p_in_game = set()
        for turn_data in turns.values():
            all_p_in_game.update(turn_data['players'])
        # Ensure filtering out potential Nones before sorting
        all_p_in_game = {p for p in all_p_in_game if p is not None}
        if not all_p_in_game: continue # Skip game if no valid players identified

        sorted_p = sorted(list(all_p_in_game)); p2i = {p:i for i,p in enumerate(sorted_p)}; i2p = {i:p for p,i in p2i.items()}; num_p_game=len(sorted_p)
        player_maps_game[game_id]=(p2i, i2p)

        if num_p_game == 0 or num_p_game > MAX_PLAYERS: continue # Skip if player count is invalid

        for (yr,sn),t_data in turns.items():
            tk=(game_id, yr, sn)
            curr_p = sorted([p for p in t_data['players'] if p is not None]) # Ensure current players are valid
            X = np.zeros((MAX_PLAYERS,1),dtype=np.float32); M = np.zeros(MAX_PLAYERS,dtype=bool)

            for p in curr_p:
                idx=p2i.get(p); # Look up index using game map
                if idx is not None and idx < MAX_PLAYERS: # Ensure index is within bounds
                    score=score_lookup.get((game_id,yr,sn,p), 0.0)
                    X[idx,0]=0.0 if pd.isna(score) else float(score)
                    M[idx]=True # Mark as present

            # Adjacency based on messages in this turn
            Ab=np.zeros((MAX_PLAYERS,MAX_PLAYERS),dtype=np.float32)
            for msg in t_data['messages']:
                # Get indices safely
                i_s = p2i.get(msg.get('speaker'))
                i_r = p2i.get(msg.get('recipient'))
                # Add edge only if both speaker and recipient are mapped and valid indices
                if i_s is not None and i_r is not None and i_s < MAX_PLAYERS and i_r < MAX_PLAYERS and i_s != i_r:
                     Ab[i_s, i_r]=1.0; Ab[i_r, i_s]=1.0 # Undirected edge

            Asl = Ab+np.identity(MAX_PLAYERS,dtype=np.float32) # Adjacency + Self-loops
            graph_lookup[tk] = {'X':X,'A':Asl,'M':M}

    # Fill any potential missing turn keys with default graph if needed by generator robustness
    # This usually isn't necessary if data processing is clean.
    print(f"Graph lookup built:{len(graph_lookup)} turns.");
    return graph_lookup, player_maps_game

# --- Module 1: Custom GCN Layer ---
class GCNLayer(Layer):
    """ Basic Graph Convolutional Layer (A * X * W + b) """
    def __init__(self, units, activation=None, name=None, **kwargs):
        super().__init__(name=name, **kwargs); self.units = units; self.activation = tf.keras.activations.get(activation)
    def build(self, input_shape):
        node_feature_shape, _ = input_shape # Expecting [X_shape, A_shape]
        node_dim = node_feature_shape[-1]
        self.w = self.add_weight(shape=(node_dim, self.units), initializer='glorot_uniform', name='w', trainable=True)
        self.b = self.add_weight(shape=(self.units,), initializer='zeros', name='b', trainable=True)
        super().build(input_shape) # Call parent build method
    def call(self, inputs):
        node_features, adj_matrix = inputs; support = tf.matmul(node_features, self.w); output = tf.matmul(adj_matrix, support); output = output + self.b;
        if self.activation is not None: output = self.activation(output)
        return output
    def get_config(self):
        config = super().get_config(); config.update({"units": self.units, "activation": tf.keras.activations.serialize(self.activation)}); return config
    @classmethod
    def from_config(cls, config):
        activation_config = config.pop("activation", None) # Extract activation config if present
        activation = tf.keras.activations.deserialize(activation_config)
        config['activation'] = activation # Add deserialized activation back
        return cls(**config)


# --- Module 2: Text + Delta Feature Encoder ---
class TextFeatureEncoder(Model):
    """ Encodes text sequence and delta features into a combined embedding. """
    def __init__(self, max_vocab_size, embedding_dim, seq_len, num_heads, dropout_rate, delta_feature_dim, name="text_feature_encoder", **kwargs):
        super().__init__(name=name, **kwargs); self.embedding=Embedding(input_dim=max_vocab_size, output_dim=embedding_dim, name='embedding'); self.mha=MultiHeadAttention(num_heads=num_heads, key_dim=embedding_dim, name='attention'); self.add_norm1=Add(name='add_resid_1'); self.layernorm1=LayerNormalization(epsilon=1e-6, name='norm_1'); self.pooling=GlobalAveragePooling1D(name='text_pooling'); self.dropout_text1=Dropout(dropout_rate, name='text_dropout_1'); self.dense_features=Dense(16, activation='relu', name='dense_delta_features'); self.concatenate=Concatenate(name='concat_text_features'); self.dropout_combined=Dropout(dropout_rate*1.5, name='text_feature_dropout'); self.output_dim = embedding_dim; self.output_dense=Dense(self.output_dim, activation='relu', name='text_feature_output') # Ensure output dim defined
    def call(self, inputs, training=False): text_input, feature_input = inputs; x_txt = self.embedding(text_input); attn_output = self.mha(query=x_txt, value=x_txt, key=x_txt); x_txt = self.add_norm1([x_txt, attn_output]); x_txt = self.layernorm1(x_txt); text_emb = self.pooling(x_txt); text_emb = self.dropout_text1(text_emb, training=training); feature_proc = self.dense_features(feature_input); combined = self.concatenate([text_emb, feature_proc]); combined_dropout = self.dropout_combined(combined, training=training); output_embedding = self.output_dense(combined_dropout); return output_embedding
    def get_config(self): config=super().get_config(); config.update({'max_vocab_size':self.embedding.input_dim, 'embedding_dim':self.embedding.output_dim, 'seq_len':MAX_SEQUENCE_LENGTH, 'num_heads':self.mha.num_heads,'dropout_rate':self.dropout_text1.rate, 'delta_feature_dim':DELTA_FEATURE_DIM}); return config # Example parameters
    @classmethod
    def from_config(cls, config): return cls(**config)


# --- Module 3: Graph Encoder ---
class GraphEncoder(Model):
    """ Encodes graph structure and node features using custom GCN layers. """
    def __init__(self, max_players, node_feature_dim, gcn_units, dropout_rate, final_emb_dim, name="graph_encoder", **kwargs):
        super().__init__(name=name, **kwargs); self.max_players=max_players; self.final_emb_dim=final_emb_dim
        self.gcn1=GCNLayer(gcn_units, activation='relu', name='CustomGCN1'); self.dropout_gcn1=Dropout(dropout_rate)
        self.gcn2=GCNLayer(gcn_units, activation='relu', name='CustomGCN2'); self.dropout_gcn2=Dropout(dropout_rate)
        self.gcn3=GCNLayer(gcn_units // 2, activation='relu', name='CustomGCN3')
        self.graph_pooling=GlobalAveragePooling1D(name='GraphPooling'); self.dropout_pooled=Dropout(dropout_rate * 1.5)
        self.output_dense=Dense(self.final_emb_dim, activation='relu', name='graph_output')
    def call(self, inputs, training=False):
        node_features, adj_matrix, node_mask = inputs; node_mask_float=tf.cast(node_mask,dtype=tf.float32); node_mask_expanded=tf.expand_dims(node_mask_float,axis=-1); masked_node_features=node_features*node_mask_expanded
        x_graph = self.gcn1([masked_node_features, adj_matrix]); x_graph = self.dropout_gcn1(x_graph, training=training)
        x_graph = self.gcn2([x_graph, adj_matrix]); x_graph = self.dropout_gcn2(x_graph, training=training)
        x_graph = self.gcn3([x_graph, adj_matrix]);
        x_graph._keras_mask = node_mask # Set mask before pooling
        graph_emb_pooled = self.graph_pooling(x_graph); graph_emb_pooled = self.dropout_pooled(graph_emb_pooled, training=training); graph_embedding = self.output_dense(graph_emb_pooled)
        return graph_embedding
    def get_config(self): config = super().get_config(); config.update({'max_players':self.max_players,'node_feature_dim':1,'gcn_units':self.gcn1.units,'dropout_rate':self.dropout_gcn1.rate,'final_emb_dim':self.final_emb_dim}); return config
    @classmethod
    def from_config(cls, config): return cls(**config)

# --- Module 4: Main Hybrid Model ---
class DiplomacyHybridModel(Model):
    """ Combines text/feature and graph encoders using attention for classification. """
    def __init__(self, text_feature_encoder, graph_encoder, reason_dropout_rate, attention_heads, name="diplomacy_hybrid", **kwargs):
        super().__init__(name=name, **kwargs); self.text_feature_encoder = text_feature_encoder; self.graph_encoder = graph_encoder; self.concatenate = Concatenate(name='combine_embeddings')
        # Calculate combined_dim dynamically based on sub-module output shapes IF THEY ARE BUILT
        # This requires the sub-models to be built *before* the parent model __init__ runs fully,
        # which is tricky. Easier to define expected dims or pass them in.
        # Hardcoding based on current definition: text=EMBEDDING_DIM, graph=EMBEDDING_DIM//2
        combined_dim = EMBEDDING_DIM + (EMBEDDING_DIM // 2)
        self.reason_dense1 = Dense(combined_dim, activation='relu', name='reason_dense1'); self.dropout_reason1 = Dropout(reason_dropout_rate)
        self.reshape_for_attn = Reshape((1, combined_dim)); attn_key_dim = max(16, combined_dim // attention_heads); self.attention_reason = MultiHeadAttention(num_heads=attention_heads, key_dim=attn_key_dim, name='reason_attention'); self.reshape_after_attn = Reshape((combined_dim,)); self.dropout_reason2 = Dropout(reason_dropout_rate)
        self.output_classifier = Dense(1, activation='sigmoid', name='final_output')
    def call(self, inputs, training=False):
        text_input, feature_input, node_features, adj_matrix, node_mask = inputs
        text_feat_emb = self.text_feature_encoder([text_input, feature_input], training=training)
        graph_emb = self.graph_encoder([node_features, adj_matrix, node_mask], training=training)
        combined = self.concatenate([text_feat_emb, graph_emb])
        x = self.reason_dense1(combined); x = self.dropout_reason1(x, training=training)
        x_reshaped = self.reshape_for_attn(x); attn_output = self.attention_reason(query=x_reshaped, value=x_reshaped, key=x_reshaped, training=training); x = self.reshape_after_attn(attn_output); x = self.dropout_reason2(x, training=training)
        output = self.output_classifier(x)
        return output
    # get_config / from_config important for saving model correctly
    def get_config(self):
        # Serialize sub-modules by getting their configs
        # This relies on submodules having proper get_config implemented
        config = super().get_config()
        config.update({
            "text_feature_encoder": tf.keras.layers.serialize(self.text_feature_encoder),
            "graph_encoder": tf.keras.layers.serialize(self.graph_encoder),
            "reason_dropout_rate": self.dropout_reason1.rate,
            "attention_heads": self.attention_reason.num_heads
         })
        return config
    @classmethod
    def from_config(cls, config):
        # Deserialize sub-models before initializing parent
        config["text_feature_encoder"] = tf.keras.layers.deserialize(config["text_feature_encoder"])
        config["graph_encoder"] = tf.keras.layers.deserialize(config["graph_encoder"])
        return cls(**config)


# --- Main Script ---
print("--- Starting Hybrid GCN + Transformer (Modular, Downsampling, Predefined Splits, Corrected) ---")

# 1. Load Data from Separate Files
print("--- Loading Train/Val/Test Data ---")
train_raw = load_jsonl_dataset(os.path.join(DATA_DIR, 'train.jsonl'))
val_raw   = load_jsonl_dataset(os.path.join(DATA_DIR, 'validation.jsonl'))
test_raw  = load_jsonl_dataset(os.path.join(DATA_DIR, 'test.jsonl'))
if train_raw is None or val_raw is None or test_raw is None: raise ValueError("Failed to load data files.")
all_raw_data = train_raw + val_raw + test_raw # For lookups/tokenizer

# 2. Preprocess ALL Data for Lookups
print("\n--- Preprocessing ALL Data for Lookups ---")
all_processed = preprocess_for_hybrid(all_raw_data);
if not all_processed: raise ValueError("Preprocessing failed.")
score_lookup = create_score_lookup(all_processed)
games_data_agg = aggregate_data_by_turn(all_processed)
graph_lookup, _ = build_turn_graph_lookup(games_data_agg, score_lookup)

# 3. Prepare SEPARATE Data Splits
print("\n--- Preparing Train/Val/Test Data Splits ---")
train_processed = preprocess_for_hybrid(train_raw); val_processed = preprocess_for_hybrid(val_raw); test_processed = preprocess_for_hybrid(test_raw)
if not train_processed or not val_processed or not test_processed: raise ValueError("Data splits empty after preprocessing.")

# 4. Fit Tokenizer (on ALL text) and Transform Splits
print("\n--- Tokenizing Text Data ---")
all_texts = [s['clean_text'] for s in all_processed]
tokenizer = Tokenizer(num_words=MAX_VOCAB_SIZE, oov_token='<OOV>')
tokenizer.fit_on_texts(all_texts); vocab_size = min(MAX_VOCAB_SIZE, len(tokenizer.word_index) + 1)
print(f"Vocab size: {vocab_size}")
def tokenize_and_pad(samples): texts=[s['clean_text'] for s in samples]; sequences=tokenizer.texts_to_sequences(texts); return pad_sequences(sequences,maxlen=MAX_SEQUENCE_LENGTH,padding='post',truncating='post')
padded_train=tokenize_and_pad(train_processed); padded_val=tokenize_and_pad(val_processed); padded_test=tokenize_and_pad(test_processed)
print(f"Padded Shapes: Tr={padded_train.shape}, Vl={padded_val.shape}, Ts={padded_test.shape}")

# 5. Prepare Other Features/Labels for Splits
def extract_features_labels_keys(samples): labels=np.array([s['label'] for s in samples]); delta_feats=np.array([extract_delta_features(s, score_lookup) for s in samples]); turn_keys=[(s['game_id'],s['year'],s['season']) for s in samples]; return labels, delta_feats, turn_keys
labels_train, delta_feats_train, turn_keys_train = extract_features_labels_keys(train_processed)
labels_val, delta_feats_val, turn_keys_val = extract_features_labels_keys(val_processed)
labels_test, delta_feats_test, turn_keys_test = extract_features_labels_keys(test_processed)

# 6. Apply Downsampling (ONLY on TRAINING data)
print("\n--- Downsampling Training Data ---")
train_indices = np.arange(len(labels_train)); minority_indices = train_indices[labels_train == 0]; majority_indices = train_indices[labels_train == 1]
print(f"Original Train counts: Lie(0)={len(minority_indices)}, Truth(1)={len(majority_indices)}")
if len(minority_indices)==0 or len(majority_indices)==0: raise ValueError("Cannot downsample, one class empty.")
majority_downsampled_indices = resample(majority_indices, replace=False, n_samples=len(minority_indices), random_state=SEED)
downsampled_train_indices = np.concatenate([minority_indices, majority_downsampled_indices]); np.random.shuffle(downsampled_train_indices)
padded_train_ds = padded_train[downsampled_train_indices]; delta_feats_train_ds = delta_feats_train[downsampled_train_indices]; labels_train_ds = labels_train[downsampled_train_indices]
turn_keys_train_list = [turn_keys_train[i] for i in downsampled_train_indices]
print(f"Downsampled Train counts: Lie(0)={sum(labels_train_ds==0)}, Truth(1)={sum(labels_train_ds==1)}")

# 7. Create tf.data.Datasets
print("\n--- Creating tf.data Datasets ---")
def get_graph_data_batch(keys): # Needs definition here
    batch_X, batch_A, batch_M = [],[],[]; default_graph={'X':np.zeros((MAX_PLAYERS,1),dtype=np.float32),'A':np.identity(MAX_PLAYERS,dtype=np.float32),'M':np.zeros(MAX_PLAYERS,dtype=bool)};
    for key in keys: graph_data=graph_lookup.get(key,default_graph); batch_X.append(graph_data['X']); batch_A.append(graph_data['A']); batch_M.append(graph_data['M']);
    return np.array(batch_X),np.array(batch_A),np.array(batch_M)
def data_generator_split(text_padded_arr, delta_feats_arr, labels_arr, turn_keys_list): # Ensure defined before use
    num_samples = len(labels_arr); indices = np.arange(num_samples)
    for i in indices: text_pad=text_padded_arr[i]; delta_f=delta_feats_arr[i]; label=labels_arr[i]; turn_key=turn_keys_list[i]; X_g_batch, A_g_batch, M_g_batch = get_graph_data_batch([turn_key]); yield (text_pad, delta_f, X_g_batch[0], A_g_batch[0], M_g_batch[0]), label
output_signature=((tf.TensorSpec(shape=(MAX_SEQUENCE_LENGTH,),dtype=tf.int32), tf.TensorSpec(shape=(DELTA_FEATURE_DIM,),dtype=tf.float32), tf.TensorSpec(shape=(MAX_PLAYERS,1),dtype=tf.float32), tf.TensorSpec(shape=(MAX_PLAYERS,MAX_PLAYERS),dtype=tf.float32), tf.TensorSpec(shape=(MAX_PLAYERS,),dtype=tf.bool)), tf.TensorSpec(shape=(),dtype=tf.int32))
# Create datasets
train_dataset=tf.data.Dataset.from_generator(lambda:data_generator_split(padded_train_ds,delta_feats_train_ds,labels_train_ds,turn_keys_train_list),output_signature=output_signature).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE).repeat()
val_dataset=tf.data.Dataset.from_generator(lambda:data_generator_split(padded_val,delta_feats_val,labels_val,turn_keys_val),output_signature=output_signature).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
test_dataset=tf.data.Dataset.from_generator(lambda:data_generator_split(padded_test,delta_feats_test,labels_test,turn_keys_test),output_signature=output_signature).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE); print("tf.data Datasets created.")

# 8. Build & Compile the MAIN Hybrid Model
print("\n--- Building and Compiling Main Hybrid Model ---")
text_encoder = TextFeatureEncoder(max_vocab_size=vocab_size, embedding_dim=EMBEDDING_DIM, seq_len=MAX_SEQUENCE_LENGTH,num_heads=NUM_HEADS_TXT, dropout_rate=DROPOUT_RATE, delta_feature_dim=DELTA_FEATURE_DIM)
graph_encoder = GraphEncoder(max_players=MAX_PLAYERS, node_feature_dim=1, gcn_units=GCN_UNITS,dropout_rate=DROPOUT_RATE, final_emb_dim=EMBEDDING_DIM // 2)
diplomacy_model = DiplomacyHybridModel(text_feature_encoder=text_encoder, graph_encoder=graph_encoder, reason_dropout_rate=0.3, attention_heads=NUM_HEADS_REASON)
diplomacy_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE), loss='binary_crossentropy', metrics=['accuracy']) # Compile with basic metrics
# Build model explicitly by calling it once
input_specs = [spec for spec in output_signature[0]]
# Use tf.keras.Input to define symbolic inputs based on specs
symbolic_inputs = [tf.keras.Input(batch_shape=(None,) + spec.shape[1:], dtype=spec.dtype) if spec.shape else tf.keras.Input(batch_shape=(None,), dtype=spec.dtype) for spec in input_specs ] # Handle scalar shape for label spec (though label is output) - need input specs only
input_shapes_only = [spec.shape for spec in output_signature[0]] # Shapes of the 5 inputs
diplomacy_model.build(input_shape=input_shapes_only) # Build expects list of shapes
diplomacy_model.summary(expand_nested=True)

# 9. Train the Main Hybrid Model
model_name_main_hybrid="DiplomacyModularHybrid_DS_Attn"; main_hybrid_filename=model_name_main_hybrid.lower()+".keras"; main_hybrid_path=os.path.join(OUTPUT_DIR, f"best_{main_hybrid_filename}")
main_early_stopping=EarlyStopping(monitor='val_loss',patience=PATIENCE,verbose=1,restore_best_weights=True,mode='min'); main_checkpoint=ModelCheckpoint(main_hybrid_path,monitor='val_loss',save_best_only=True,mode='min',verbose=1)
print(f"\n--- Training Main Hybrid Model: {model_name_main_hybrid} ---");
steps_per_epoch_train = len(downsampled_train_indices) // BATCH_SIZE
if len(downsampled_train_indices) % BATCH_SIZE != 0: steps_per_epoch_train += 1
history_main=diplomacy_model.fit(train_dataset,validation_data=val_dataset,epochs=EPOCHS,steps_per_epoch=steps_per_epoch_train,callbacks=[main_early_stopping, main_checkpoint],verbose=1)

# 10. Evaluate the Main Hybrid Model...
print(f"\n--- Evaluating Main Hybrid Model on Test Set ---")
if os.path.exists(main_hybrid_path): diplomacy_model.load_weights(main_hybrid_path); print("Loaded best main hybrid weights.")
else: print("Warn: Best main hybrid model file not found.")
test_loss_mh, test_acc_mh = diplomacy_model.evaluate(test_dataset, verbose=1)
print("Generating predictions for detailed metrics...")
y_pred_probs_mh=diplomacy_model.predict(test_dataset); y_pred_bin_mh=(y_pred_probs_mh > 0.5).astype(int).flatten(); y_true_mh = labels_test
f1_lie_mh=f1_score(y_true_mh, y_pred_bin_mh, pos_label=0, zero_division=0); f1_truth_mh=f1_score(y_true_mh, y_pred_bin_mh, pos_label=1, zero_division=0); f1_macro_mh=f1_score(y_true_mh, y_pred_bin_mh, average='macro', zero_division=0); f1_weighted_mh=f1_score(y_true_mh, y_pred_bin_mh, average='weighted', zero_division=0)
print("\nMain Hybrid Model Test Results (Using Sklearn Metrics):"); print(f" Accuracy:{test_acc_mh:.4f}, Loss:{test_loss_mh:.4f}"); print(f" Macro F1:{f1_macro_mh:.4f}, Weighted F1:{f1_weighted_mh:.4f}"); print("-" * 20); print(classification_report(y_true_mh, y_pred_bin_mh, target_names=['Lie (0)', 'Truth (1)'], digits=4)); print("-" * 20)

# 11. Plot Confusion Matrix...
conf_matrix_mh=confusion_matrix(y_true_mh, y_pred_bin_mh); plt.figure(figsize=(6,5)); sns.heatmap(conf_matrix_mh, annot=True, fmt='d', cmap='Blues', xticklabels=['Pred Lie(0)','Pred Truth(1)'], yticklabels=['Actual Lie(0)','Actual Truth(1)']); plt.title(f'Hybrid Model (DS, Attn) CM\nAcc:{test_acc_mh:.3f}|MacroF1:{f1_macro_mh:.3f}|Lie F1:{f1_lie_mh:.3f}',fontsize=9); plt.ylabel('Actual'); plt.xlabel('Predicted'); plt.tight_layout(); plot_cm_mh_path = os.path.join(OUTPUT_DIR, f"cm_main_hybrid_ds_attn.png");
try: plt.savefig(plot_cm_mh_path); print(f"Saved CM plot: {plot_cm_mh_path}")
except Exception as e: print(f"Error saving plot: {e}")
plt.show()

print("\n--- Script Finished ---")
```
